### GroupBy()

If someone asks me:

"What's the one Pandas feature every data analyst must know?"

My answer would be:

groupby()

Almost every real dataset analysis uses it.

🤔 What is groupby()?

Imagine you have this employee data:
| Name  | Department | Salary |
| ----- | ---------- | -----: |
| Rahul | IT         |  45000 |
| Neha  | IT         |  60000 |
| Aditi | HR         |  50000 |
| Aman  | HR         |  55000 |
| Priya | Finance    |  70000 |

Suppose your manager asks:

"What is the average salary of each department?"

Without groupby(), you would have to manually separate IT, HR, Finance...

With groupby(), Pandas does it automatically.

Think of it like this

groupby() works in 3 steps:

Data
   ->
Group by some column
   ->
Apply a calculation

For example:

Employees
      ->
Group by Department
      ->
Find average salary

Syntax
df.groupby("Column_Name")

This only creates groups.

It doesn't calculate anything yet.

You must tell Pandas what you want.

Example:

df.groupby("Department")["Salary"].mean()

Read it like English:

Group employees by Department, then calculate the mean Salary.

```python
import pandas as pd

employee = {
    "Name": ["Rahul","Aditi","Neha","Aman","Priya","Rohan","Sneha","Karan"],
    "Department": ["IT","HR","IT","Finance","HR","Finance","IT","HR"],
    "Salary": [45000,52000,60000,70000,55000,65000,58000,50000],
    "Experience": [2,4,5,6,3,7,4,2]
}

df = pd.DataFrame(employee)
print(df)
print(df.groupby("Department"))
#1️⃣ Average Salary by Department
print(
    df.groupby("Department")["Salary"].mean()
)

#2️⃣ Maximum Salary by Department
print(
    df.groupby("Department")["Salary"].max()
)

#3️⃣ Minimum Salary
print(
    df.groupby("Department")["Salary"].min()
)

#4️⃣ Total Salary
print(
    df.groupby("Department")["Salary"].sum()
)
#5️⃣ Count Employees
print(
    df.groupby("Department")["Name"].count()
)

#6️⃣ Multiple Calculations at Once
print(
    df.groupby("Department")["Salary"].agg(
        ["mean", "max", "min", "sum", "count"]
    )
)
```

### 📝 Practice Questions

```python
import pandas as pd
employee = {
    "Name": ["Rahul","Aditi","Neha","Aman","Priya","Rohan","Sneha","Karan"],
    "Department": ["IT","HR","IT","Finance","HR","Finance","IT","HR"],
    "Salary": [45000,52000,60000,70000,55000,65000,58000,50000],
    "Experience": [2,4,5,6,3,7,4,2]
}
df = pd.DataFrame(employee)
print(df)

#Q1 Find the average salary of each department.
print(df.groupby("Department")['Salary'].mean())

#Q2 Find the highest salary in each department.
print(df.groupby('Department')['Salary'].max())

#Q3 Find the lowest salary in each department.
print(df.groupby("Department")['Salary'].min())

#Q4 Find the total salary paid in each department.
print(df.groupby("Department")['Salary'].sum())

#Q5 Count how many employees are in each department.
print(df.groupby("Department")["Salary"].count())

#Q6 Find the average experience of each department.
print(df.groupby("Department")["Experience"].mean())

#Q7 Display all salary statistics together using: agg()
print(df.groupby("Department")["Salary"].agg(['mean','max','min','sum','count']))

#Q8 Which department has the highest average salary?
print(df.groupby("Department")["Salary"].mean().sort_values(ascending=False))

#Q9 Which department has the highest total salary payout?
print(df.groupby("Department")["Salary"].sum().sort_values(ascending=False))
```

### Advanced GroupBy

```python
import pandas as pd

df = pd.read_csv("employee_data.csv")

#1️⃣ Group By Multiple Columns
print(
    df.groupby(["Department", "City"])["Salary"].mean()
)

# #2️⃣ Understanding MultiIndex

# #The output looks like this:
# # Department
# #       ↓
# # IT
# #       ↓
# # Pune
# #       ↓
# # 46000
# Pandas first groups by Department, then within each department it groups by City.

# Think of it like folders:
# Department
# │
# ├── IT
# │     ├── Pune
# │     ├── Bangalore
# │
# ├── HR
# │     ├── Mumbai
# │     ├── Pune
# │
# └── Finance
#       ├── Delhi
#       ├── Mumbai

#3️⃣ Multiple Columns with agg()
print(
    df.groupby("Department").agg({
        "Salary": ["mean", "max"],
        "Experience": ["mean", "max"]
    })
)
```

### reset_index()

```python
#When you use groupby(), Pandas makes the grouped column the index.

#Without using reset_index()
print(df1.groupby("Department")["Salary"].mean())

#With using reset_index()
print(
    df1.groupby("Department")["Salary"].mean().reset_index()
)

# Notice:

# 👉 Department is not a normal column anymore.

# It has become the index.

# Sometimes we don't want that so we use reset_index()
```

### 📝 Practice Questions

```python
#Q1 Find the average salary by department without using reset_index().
print(df1.groupby("Department")["Salary"].mean())

#Q2 Find the average salary by department with reset_index().
print(df1.groupby("Department")["Salary"].mean().reset_index())

#Q3 Find the average salary by Department and City, then convert the result into a normal DataFrame using reset_index()
avg = df1.groupby(["Department","City"])['Salary'].mean()
print(avg.reset_index())

#Q4 Find the maximum experience for each department and return it as a normal DataFrame.
max = df1.groupby("Department")['Experience'].max()
print(max.reset_index())

# as_index=False
#  tells Pandas not to make the grouped column an index in the first place.
# Many professionals prefer as_index=False because it saves an extra step.

#Find the average salary by department using as_index=False.
print(df.groupby("Department",as_index=False)['Salary'].mean())

#Find the total salary by department using as_index=False.
print(df.groupby("Department",as_index=False)['Salary'].sum())

#Find the maximum experience by department using as_index=False.
print(df.groupby("Department",as_index=False)['Experience'].max())

#Find the average salary by Department and City using as_index=False.
print(df1.groupby(["Department","City"],as_index=False)['Salary'].mean())

#Create a summary table showing: Salary → mean, max, Experience → mean Use as_index=False.
print(df.groupby('Department',as_index=False).agg({
    "Salary" : ['mean','max'],
    "Experience" : ['mean']
}))
```

### size() count() nunique()

```python
import pandas as pd
df = pd.read_csv('orders_data.csv')
print(df)

print(df.groupby('Category').size())  #size() counts every row, even if some values are missing.
print(df.groupby("Category").count())  #Because count() ignores NaN values.
print(df.groupby("Category")["Customer"].nunique()) #gives only unique value count no repeated
```

### Practice Questions

```python
import pandas as pd
df = pd.read_csv('orders_data.csv')
df1 = df.head()
print(df1)

#Q1 Count how many orders each Category has using: size()
print(df1.groupby("Category",as_index=False).size())

#Q2 Count how many orders each Category has using: count()
print(df1.groupby("Category",as_index=False)['Order_ID'].count())

#Q3 Count how many orders each City has using size().
print(df1.groupby("City",as_index=False)['Category'].size())

#Q4 Count how many orders each Payment_Method has using count().
df1.groupby("Payment_Method", as_index=False)["Order_ID"].count()

#Q5 Find the average Price of each Category.
print(df1.groupby("Category",as_index=False)['Price'].mean())

#Q6 Find the highest Discount given in each Category.
print(df1.groupby("Category",as_index=False)['Discount'].max())

#Q7 Find: mean Price, max Price, min Price, mean Rating for every Category.
print(df1.groupby("Category",as_index=False).agg({
    "Price" : ['mean','max','min'],
    "Rating" : ['mean']
}))

#Q8 Which City has the highest average order price?
avg = df1.groupby("City")['Price'].mean()
print(avg.sort_values(ascending=False))

#Q9 Find the number of unique customers in each category.
print(df.groupby('Category')['Customer'].nunique())

#Q10 Find the number of unique products in each city.
print(df.groupby('City')['Product'].nunique())

#Q11 Find the number of unique payment methods used in each city.
print(df.groupby("City")['Payment_Method'].nunique())

#Q12 Find: Total Orders (size) Unique Customers (nunique) for every category.
print(df.groupby('Category').agg({
    "Order_ID" : ['size'],
    "Customer" : ['nunique']
}))

#Q13 Create one summary table using agg() that shows:Price → mean Rating → mean Customer → nunique
print(df.groupby("Category").agg({
    "Price" : ['mean'],
    "Rating" : ['mean'],
    "Customer" : ['nunique'] 
}))
```

| Function    | Counts          |

### value_counts()

```python
import pandas as pd
# This is another function you'll use all the time.

# Suppose your manager asks:

# "How many orders came from each city?"

# There are two ways to answer:

# Method 1: groupby()
# df.groupby("City").size()
# Method 2: value_counts()
# df["City"].value_counts()

# Both count occurrences, but value_counts() is much shorter and is the preferred choice when you only need counts for a single column.

df = pd.read_csv("orders_data.csv")

#Q1 Count how many orders came from each city using value_counts()
print(df["City"].value_counts())

#Q2 Count how many orders belong to each category.
print(df['Category'].value_counts())

#Q3 Count how many orders were made using each payment method.
print(df['Payment_Method'].value_counts())

#Q4 Which customer placed the most orders?
print(df['Customer'].value_counts().head(1))

#Q5 Which product was ordered the most?
print(df['Product'].value_counts().head(1))
print(df["Product"].value_counts().idxmax())
```

📚 When should I use groupby() vs value_counts()?

Use value_counts() when you're counting values in one column.
1. df["City"].value_counts()
2. df["Category"].value_counts()

Use groupby() when you're calculating something for each group.
1. df.groupby("Category")["Price"].mean()
2. df.groupby("City")["Discount"].sum()
3. df.groupby("Department")["Salary"].max()
