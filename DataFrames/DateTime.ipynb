### 📅 1. Date & Time (.dt) ⭐⭐⭐⭐⭐

This is one of the most frequently used topics in real-world datasets.

You'll learn:

1. pd.to_datetime()
2. .dt.year
3. .dt.month
4. .dt.day
5. .dt.day_name()
6. .dt.month_name()
7. Filtering by dates
8. Calculating date differences

```python
import pandas as pd

employee = {
    "Name": ["Rahul", "Neha", "Aman", "Priya", "Karan"],
    "Joining_Date": [
        "2021-03-15",
        "2020-07-10",
        "2022-01-25",
        "2019-11-05",
        "2023-06-18"
    ],
    "Salary": [50000, 60000, 70000, 65000, 55000]
}

df = pd.DataFrame(employee)
df["Joining_Date"] = pd.to_datetime(df["Joining_Date"])
print(df)
print(df.dtypes)

#1️⃣ Year
print(df["Joining_Date"].dt.year)

#2️⃣ Month
print(df["Joining_Date"].dt.month)

#3️⃣ Day
print(df["Joining_Date"].dt.day)

#4️⃣ Month Name
print(df["Joining_Date"].dt.month_name())

#5️⃣ Day Name
print(df["Joining_Date"].dt.day_name())
```

### 📝 Practice

```python
import pandas as pd
employee = {
    "Name": ["Rahul", "Neha", "Aman", "Priya", "Karan"],
    "Joining_Date": [
        "2021-03-15",
        "2020-07-10",
        "2022-01-25",
        "2019-11-05",
        "2023-06-18"
    ],
    "Salary": [50000, 60000, 70000, 65000, 55000]
}
df = pd.DataFrame(employee)

#Q1 Convert Joining_Date into datetime.
df["Joining_Date"] = pd.to_datetime(df["Joining_Date"])
print(df)

#Q2 Create a new column: Joining_Year
df["Joining_Year"] = df["Joining_Date"].dt.year

#Q3 Create a new column: Joining_Month
df["Joining_Month"] = df["Joining_Date"].dt.month

#Q4 Create a new column: Joining_Day
df["Joining_Day"] = df["Joining_Date"].dt.day

#Q5 Create a new column: Month_Name
df["Month_Name"] = df["Joining_Date"].dt.month_name()

#Q6 Create a new column: Day_Name
df["Day_Name"] = df["Joining_Date"].dt.day_name()
print(df)

#Q7 Display only employees who joined in 2021.
print(df[df["Joining_Year"] == 2021])
# condition = df[df["Joining_Year"] == 2021]
# print(condition)

#Q8 Display only employees who joined in the month of July.
print(df[df["Month_Name"] == "July"])
# con = df[df["Month_Name"] == "July"]
# print(con)

#Q9 Display only employees who joined after 2021.
print(df[df["Joining_Date"].dt.year > 2021])
```
