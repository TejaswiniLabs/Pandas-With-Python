### 📘 map() vs apply()

### 🤔 What is map()?

map() is used to change or replace values in a Series (single column).

Think of it as:

"Take every value in this column and convert it into something else."

1. Replace()

```python
import pandas as pd

#Example 1
df = pd.DataFrame({
    "Gender" : ["Male", "Female", "Female", "Male"]
})
print(df)

df["Gender"] = df["Gender"].map({
    "Male" : 0,
    "Female" : 1
})
print(df)

#Example 2
df1 = pd.DataFrame({
    "Department" : ["Information Technology", "Human Resources", "Finance"]
})
print(df1)
df1["Department"] = df1["Department"].map({
    "Information Technology" : "IT",
    "Human Resources" : "HR",
    "Finance" : "FIN"
})
print(df1)
print()

grades = pd.Series(["A", "B", "A", "C", "B"])

print(grades.map({
    "A": "Excellent",
    "B": "Good",
    "C": "Average"
}))

```

Important Rule

✅ map() works on a Series only.

This works:

df["Gender"].map(...)

❌ This does not work:

df.map(...)

because df is a DataFrame, not a Series.

### 📝 Practice

```python
import pandas as pd

employee = {
    "Name": ["Rahul", "Neha", "Aman", "Priya", "Karan"],
    "Gender": ["Male", "Female", "Male", "Female", "Male"],
    "Department": ["Information Technology",
                   "Finance",
                   "Human Resource",
                   "Information Technology",
                   "Finance"]
}

df = pd.DataFrame(employee)

print(df)

#Q1 Convert: Male → M Female → F using map().
df["Gender"] = df["Gender"].map({
    "Male" : "M",
    "Female" : "F"
})
print(df)

#Q2 Convert Department: Information Technology → IT Finance → FIN Human Resource → HR
df["Department"] = df["Department"].map({
    "Information Technology" : "IT",
    "Human Resource" : "HR",
    "Finance" : "FIN"
})
print(df)

#Q3 How many employees are in the IT department after mapping?
print(df[df["Department"] == "IT"].shape[0])

print(type(df))
print(type(df["Gender"]))
```

### 🤔 Why is it not "Both"?

Look carefully.

You wrote:

df["Gender"] = df["Gender"].map({...})

You thought:

"I'm using a DataFrame."

Actually...

No.

Let's break it down.

df is a DataFrame
print(type(df))

Output:

<class 'pandas.core.frame.DataFrame'>
But
df["Gender"]

is NOT a DataFrame.

It is a Series.

print(type(df["Gender"]))

Output:

<class 'pandas.core.series.Series'>

So this line:

df["Gender"].map(...)

means:

Take the Gender column (a Series)
Apply .map() to that Series
Store the result back into the DataFrame

So map() is never applied to the DataFrame itself.

### 📘 What is apply()?

apply() is used to apply a function to a Series or DataFrame.

Think of it like this:

1. map() → Replace values
2. apply() → Perform operations using a function

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Rahul", "Neha", "Aman"],
    "Salary": [50000, 60000, 70000]
})

print(df)

df["Salary"] = df["Salary"].apply(lambda x: x + 5000)
print("Increased salary by 5000:\n",df)

df["Name"] = df["Name"].apply(lambda a : a.upper())
print(df)

df["Salary"] = df["Salary"].apply(lambda x: x * x)
print(df)
```

### What is lambda?

This:

lambda x: x + 5000

means:

For every value x,
return x + 5000

It's a short way of writing a function.

```python
#📝 Practice

import pandas as pd

employee = {
    "Name": ["Rahul","Neha","Aman","Priya","Karan"],
    "Salary": [45000,55000,65000,70000,80000],
    "Experience": [2,4,6,5,8]
}

df = pd.DataFrame(employee)

print(df)

#Q1 Increase every salary by ₹5000 using apply().
df["Salary"] = df["Salary"].apply(lambda a : a + 5000)
print(df)

#Q2 Convert every employee name to uppercase.
df["Name"] = df["Name"].apply(lambda n : n.upper())
print(df)

#Q3 Create a new column called Experience_in_Months.
df["Experience_in_Months"] = df["Experience"].apply(lambda e : e * 12)
print(df)

#Q4 Create a new column called Salary_Category.
# Rules: Salary ≥ 70000 → "High"
# Salary ≥ 50000 and <70000 → "Medium"
# Otherwise → "Low"
# Use apply().
df["Salary_Category"] = df["Salary"].apply(
    lambda x: "High"
    if x >= 70000
    else "Medium"
    if x >= 50000
    else "Low"
)
print(df)
```

### Difference between map() and apply()

| `map()`                                      | `apply()`                                                       |
| -------------------------------------------- | --------------------------------------------------------------- |
| Works on a **Series**                        | Works on a **Series** and can also be used with a **DataFrame** |
| Used for replacing or mapping values         | Used for applying custom functions or calculations              |
| Usually uses a dictionary or simple function | Uses a function (`lambda` or a normal function)                 |
| Best for simple value transformations        | Best for complex transformations                                |
