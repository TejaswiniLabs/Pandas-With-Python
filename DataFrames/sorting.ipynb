### Sorting DataFrames

1️⃣ Sorting a Single Column

```python
import pandas as pd

df = pd.read_csv('employee_data.csv')
#print(df)

print(df.sort_values('Salary')) #lowest to highest
df.sort_values("Salary", ascending=False) #highest to lowest

#Sort by Age
print(df.sort_values('Age'))

#Sort by Name
print(df.sort_values('Name'))

```

2️⃣ Sorting Multiple Columns

```python
print(df.sort_values(by=['Department','Salary']))

#Different Order for Different Columns
print(df.sort_values(
    by=["Department", "Salary"],
    ascending=[True, False]
))
```

4️⃣ Top N Values

```python
#Top 5 Salaries
print(df.nlargest(5, "Salary"))

#Bottom 5 Salaries
print(df.nsmallest(5,'Salary'))

# Why use nlargest()?

# Instead of:

# df.sort_values(
#     "Salary",
#     ascending=False
# ).head(5)

# you can write:

# df.nlargest(5, "Salary")

# Cleaner and faster.
```

### 📝 Practice Questions

```python
import pandas as pd

df = pd.read_csv('employee_data.csv')
#print(df)

#Q1 Sort employees by Salary in ascending order.
print(df.sort_values('Salary'))

#Q2 Sort employees by Salary in descending order.
print(df.sort_values('Salary', ascending=False))

#Q3 Sort employees by Age in ascending order.
print(df.sort_values('Age'))

#Q4 Sort employees by Name alphabetically.
print(df.sort_values('Name'))

#Q5 Display the top 5 highest-paid employees.
print(df.nlargest(5, 'Salary'))

#Q6 Display the bottom 5 lowest-paid employees.
print(df.nsmallest(5, 'Salary'))

#Q7 Display only: ["Name", "Salary"] for the top 3 highest-paid employees.
print(
    df.nlargest(3, "Salary")[["Name", "Salary"]]
) 

#Q8 Sort employees by: Department (A-Z) Salary (Highest first)
print(df.sort_values(by=['Department','Salary'], ascending=[True,False]))

#Q9 Show:Name, Department, Salary for employees earning more than ₹60,000, sorted from highest salary to lowest salary.
print(df.loc[
    (df['Salary'] > 60000),
    ['Name','Department','Salary']
].sort_values('Salary', ascending=False))

#Q10 Show the top 5 employees based on:Performance_Rating Display only:Name, Department, Performance_Rating
print(
    df.nlargest(5, "Performance_Rating")
    [["Name", "Department", "Performance_Rating"]]
)

#Q11 The company wants to reward the top performers.
# Conditions:
# Salary > 60,000
# Performance_Rating > 4.5
# Display:
# Name
# Department
# Salary
# Performance_Rating
# Sort by: Performance_Rating (highest first), Salary (highest first)
print(df.loc[
    (df['Salary'] > 60000) &
    (df['Performance_Rating'] > 4.5),
    ['Name','Department','Salary','Performance_Rating']
].sort_values(by=['Performance_Rating','Salary'], ascending=[False,False]))
```
