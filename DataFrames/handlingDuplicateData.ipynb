### Handling Duplicate Data

🤔 What are Duplicates?

Sometimes the same record appears more than once.

Example:

Name	Department	Salary
Rahul	IT	45000
Aditi	HR	52000
Rahul	IT	45000
Neha	Finance	68000
Aditi	HR	52000

Here,

Rahul appears twice.
Aditi appears twice.

These are duplicate rows.

1️⃣ Create a Sample DataFrame

```python
import pandas as pd

employee = {
    "Name": [
        "Rahul",
        "Aditi",
        "Rahul",
        "Neha",
        "Aditi",
        "Aman"
    ],
    "Department": [
        "IT",
        "HR",
        "IT",
        "Finance",
        "HR",
        "IT"
    ],
    "Salary": [
        45000,
        52000,
        45000,
        68000,
        52000,
        40000
    ]
}

df = pd.DataFrame(employee)

print(df)

#2️⃣ Find Duplicate Rows
print(df.duplicated())

#3️⃣ Count Duplicate Rows
print(df.duplicated().sum())

#4️⃣ Show Only Duplicate Rows
print(df[df.duplicated()])

#5️⃣ Remove Duplicates
print(df.drop_duplicates())

#6️⃣ Remove Duplicates Permanently
print(df.drop_duplicates(inplace=True))

#7️⃣ Duplicate Based on One Column
employee = {
    "Name": ["Rahul","Rahul","Rahul","Neha"],
    "Salary":[45000,50000,55000,60000]
}
df = pd.DataFrame(employee)
print(df.duplicated(subset=['Name']))

#8️⃣ Remove Duplicates Using One Column
print(df.drop_duplicates(subset=['Name']))

#9️⃣ Keep the Last Duplicate
print(df.drop_duplicates(keep="last")) #If you want to keep the last occurrence
```

🔥 Summary

| Function              | Purpose                                 |
| --------------------- | --------------------------------------- |
| `duplicated()`        | Shows duplicate rows                    |
| `duplicated().sum()`  | Counts duplicates                       |
| `df[df.duplicated()]` | Displays duplicate rows                 |
| `drop_duplicates()`   | Removes duplicates                      |
| `subset=`             | Check duplicates using selected columns |
| `keep="last"`         | Keeps the last occurrence               |


### 📝 Practice Questions

```python
employee = {
    "Name": ["Rahul","Gandhi","Rahul","Neha","Sneha","Yogesh","Neha"],
    "Salary":[45000,55000,45000,60000,70000,80000,60000],
    "Department" : ["IT","Marketing","IT","Tech","Developer","Software","Tech"]
}
df = pd.DataFrame(employee)

#Q1 Print the DataFrame.
print(df)

#Q2 Find duplicate rows.
print(df.duplicated())

#Q3 Count duplicate rows.
print(df.duplicated().sum())

#Q4 Display only duplicate rows.
print(df[df.duplicated()])

#Q5 Remove duplicate rows.
print(df.drop_duplicates())

#Q6 Remove duplicates permanently using inplace=True.
df.drop_duplicates(inplace=True)
print(df)

#Q7 Check duplicates based only on the "Name" column.
print(df.duplicated(subset=["Name"]))

#Q8 Remove duplicates based only on "Name".
print(df.drop_duplicates(subset=["Name"]))

#Q9 
student = {
    "Name": ["Rahul","Rahul","Rahul","Neha","Neha", "Aman" ],
    "Maths": [ 80, 90, 95, 88, 91, 75 ]
}
df1 = pd.DataFrame(student)
print(df1)
print(df1.duplicated(subset=["Name"]))
print(df1[df1.duplicated(subset=["Name"])])
print(df1.drop_duplicates(subset=["Name"],keep="last"))

```
