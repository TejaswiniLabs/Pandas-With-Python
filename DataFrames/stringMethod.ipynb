### 📘 What is .str?

The .str accessor lets you perform string operations on an entire Series.

Think of it as:

"Apply string functions to every text value in a column."

For example:

df["Name"].str.upper()

converts every name to uppercase.

```python
import pandas as pd

employee = {
    "Name": ["Rahul", "  neha ", "AMAN", "Priya", "kaRan"],
    "Email": [
        "rahul@gmail.com",
        "NEHA@GMAIL.COM",
        "aman@yahoo.com",
        "priya@gmail.com",
        "karan@outlook.com"
    ],
    "Department": [
        "Information Technology",
        "Finance",
        "Human Resource",
        "Finance",
        "Information Technology"
    ]
}

df = pd.DataFrame(employee)

print(df)

#Q1 Convert every Name to uppercase.
df["Name"] = df["Name"].str.upper()
print(df)

#Q2 Convert every Email to lowercase.
df["Email"] = df["Email"].str.lower()
print(df)

#Q3 Remove extra spaces from the Name column.
df["Name"] = df["Name"].str.strip()     #Removes spaces from the beginning and end.
print(df)

#Q4 Convert every Name into proper format using:
df["Name"] = df["Name"].str.title()     #Makes the first letter of every word uppercase.
print(df)

#Q5 Create a new column called:Name_Length that stores the length of every employee's name.
df["Name_length"] = df["Name"].str.len()
print(df)

#Q6 Replace Information Technology with IT in the Department column.
df["Department"] = df["Department"].str.replace(
    "Information Technology","IT"
)
print(df)


```

### 📘 Advanced String Methods

We'll learn these five methods:

1. .str.contains()
2. .str.startswith()
3. .str.endswith()
4. .str.split()
5. .str.extract()

We'll learn them one by one.

```python
import pandas as pd

employee = {
    "Name": ["Rahul", "  neha ", "AMAN", "Priya", "kaRan"],
    "Email": [
        "rahul@gmail.com",
        "NEHA@GMAIL.COM",
        "aman@yahoo.com",
        "priya@gmail.com",
        "karan@outlook.com"
    ],
    "Department": [
        "Information Technology",
        "Finance",
        "Human Resource",
        "Finance",
        "Information Technology"
    ]
}

df = pd.DataFrame(employee)

print(df)

print(df[df["Email"].str.contains("gmail", case=False)])

print(df[df["Name"].str.startswith("R")])

print(df[df["Email"].str.endswith(".com")])

print(df["Name"].str.split())

print(df["Email"].str.extract(r"(.*)@"))
```
