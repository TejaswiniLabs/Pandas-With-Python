### 🧠 Why do we need Merge?

Imagine you're working at Amazon.

You have two CSV files.

customers.csv
| Customer_ID | Customer |
| ----------- | -------- |
| C101        | Rahul    |
| C102        | Neha     |
| C103        | Priya    |
| C104        | Aman     |
| C105        | Sneha    |

orders.csv
| Order_ID | Customer_ID | Product | Amount |
| -------- | ----------- | ------- | ------ |
| O001     | C101        | Laptop  | 65000  |
| O002     | C103        | Mobile  | 28000  |
| O003     | C102        | Shoes   | 3500   |
| O004     | C105        | Watch   | 9000   |
| O005     | C101        | Mouse   | 800    |

Question:

How do we know Rahul ordered a Laptop?

The orders.csv only has:

Customer_ID = C101

The name Rahul is stored in another table.

We need to combine the two tables.

That's exactly what merge() does.

After Merge
| Order_ID | Customer_ID | Customer | Product |
| -------- | ----------- | -------- | ------- |
| O001     | C101        | Rahul    | Laptop  |
| O002     | C103        | Priya    | Mobile  |

Now everything is in one DataFrame.

Syntax of merge:

pd.merge(
    left_dataframe,
    right_dataframe,
    on="Common_Column"
)

```python
import pandas as pd

customers = pd.DataFrame({
    "Customer_ID": ["C101", "C102", "C103", "C104"],
    "Customer": ["Rahul", "Neha", "Priya", "Aman"]
})

orders = pd.DataFrame({
    "Order_ID": ["O001", "O002", "O003", "O004"],
    "Customer_ID": ["C101", "C103", "C102", "C101"],
    "Product": ["Laptop", "Mobile", "Shoes", "Mouse"]
})

print(customers)
print()
print(orders)

# 📝 Practice Questions
#Q1 Merge the two DataFrames using Customer_ID.
result = pd.merge(customers,orders,on='Customer_ID')
print(result)

#Q2 How many rows are there after merging?
print(result.shape)

#Q3 Display only: Customer, Product from the merged DataFrame.
print(result[['Customer','Product']])

#Q4 Display only Rahul's orders.
print(result[result["Customer"] == "Rahul"])

#Q5 How many different customers placed orders?
print(result['Customer'].nunique())

```

### Types of Merge

So far we've used:
pd.merge(customers, orders, on="Customer_ID")

This is actually an Inner Join.

But there are 4 types of joins:
| Join      | Keeps                           |
| --------- | ------------------------------- |
| **Inner** | Only matching rows              |
| **Left**  | Everything from the left table  |
| **Right** | Everything from the right table |
| **Outer** | Everything from both tables     |

INNER JOIN:
Only matching rows

LEFT JOIN:
Everything from LEFT table

RIGHT JOIN:
Everything from RIGHT table

OUTER JOIN:
Everything from BOTH tables

Customers        Orders

LEFT  → Keep all Customers

RIGHT → Keep all Orders

INNER → Keep only common

OUTER → Keep everything

```python
import pandas as pd

customers = pd.DataFrame({
    "Customer_ID": ["C101", "C102", "C103", "C104", "C105"],
    "Customer": ["Rahul", "Neha", "Priya", "Aman", "Sneha"]
})

orders = pd.DataFrame({
    "Order_ID": ["O001", "O002", "O003", "O004"],
    "Customer_ID": ["C101", "C103", "C102", "C106"],
    "Product": ["Laptop", "Mobile", "Shoes", "Camera"]
})
print(customers)
print(orders)

#We can write like this also
# pd.merge(
#     customers,
#     orders,
#     left_on="Customer_ID",
#     right_on="Cust_ID"
# )

# Perform an Inner Join.
inner_join = pd.merge(customers, orders, on="Customer_ID", how="inner")
print(inner_join)

# Perform a Left Join
left_join = pd.merge(customers, orders, on="Customer_ID", how="left")
print(left_join)

#Perform a Right Join.
right_join = pd.merge(customers, orders, on="Customer_ID", how="right")
print(right_join)

#Perform an Outer Join.How many rows are there?
outer_join = pd.merge(customers, orders, on="Customer_ID", how="outer")
print(outer_join)
print(outer_join.shape)

# Which join keeps all rows from the left DataFrame?
#left join
```

### 📘 Merge with Different Column Names (left_on and right_on)

```python
import pandas as pd

customers = pd.DataFrame({
    "Customer_ID": ["C101", "C102", "C103", "C104"],
    "Customer": ["Rahul", "Neha", "Priya", "Aman"]
})

orders = pd.DataFrame({
    "Order_ID": ["O001", "O002", "O003", "O004"],
    "Cust_ID": ["C101", "C103", "C102", "C101"],
    "Product": ["Laptop", "Mobile", "Shoes", "Mouse"]
})

print(customers)
print()
print(orders)

#Merge the two DataFrames using: left_on="Customer_ID", right_on="Cust_ID" Store the result in merged_df.
merged_df = pd.merge(customers, orders, left_on="Customer_ID", right_on="Cust_ID")
print(merged_df)

#Display only: Customer, Product from the merged DataFrame.
print(merged_df[['Customer','Product']])

# Delete the Cust_ID column.
print(merged_df.drop(columns="Cust_ID", inplace=True))

# How many columns are there after deleting Cust_ID?
print(merged_df.shape)

#Display only Rahul's orders.
print(merged_df[merged_df['Customer'] == "Rahul"])
```

### 🎯 Final Merge Topic: suffixes

Sometimes both DataFrames have the same column names.

Example:

customers
| Customer_ID | Name  | City   |
| ----------- | ----- | ------ |
| C101        | Rahul | Pune   |
| C102        | Neha  | Mumbai |

orders
| Customer_ID | Product | City      |
| ----------- | ------- | --------- |
| C101        | Laptop  | Bangalore |
| C102        | Shoes   | Delhi     |

Notice both DataFrames have a City column.
If you merge them:

pd.merge(customers, orders, on="Customer_ID")

Pandas doesn't know which City is which.

So it automatically creates:
| Customer_ID | Name  | City_x | Product | City_y    |
| ----------- | ----- | ------ | ------- | --------- |
| C101        | Rahul | Pune   | Laptop  | Bangalore |

City_x → from the left DataFrame

City_y → from the right DataFrame

We can give better names:

result = pd.merge(
    customers,
    orders,
    on="Customer_ID",
    suffixes=("_Customer", "_Order")
)

print(result)

| Customer_ID | Name  | City_Customer | Product | City_Order |
| ----------- | ----- | ------------- | ------- | ---------- |
| C101        | Rahul | Pune          | Laptop  | Bangalore  |
| C102        | Neha  | Mumbai        | Shoes   | Delhi      |


```python
import pandas as pd

customers = pd.DataFrame({
    "Customer_ID": ["C101", "C102", "C103"],
    "Name": ["Rahul", "Neha", "Priya"],
    "City": ["Pune", "Mumbai", "Delhi"]
})

orders = pd.DataFrame({
    "Customer_ID": ["C101", "C102", "C103"],
    "Product": ["Laptop", "Shoes", "Mobile"],
    "City": ["Bangalore", "Hyderabad", "Chennai"]
})

#Merge both DataFrames normally.
result = pd.merge(customers, orders, on="Customer_ID")
print(result)

#Merge them again using: suffixes
suff = pd.merge(customers, orders, on="Customer_ID", suffixes=("_Customers","_Orders"))
print(suff)

#Display only:Name, City_Customer, City_Order
print(suff[["Name","City_Customers","City_Orders"]])

#How many columns are there in the merged DataFrame?
print(suff.shape)
```

### concat()

Compared to merge(), concat() is much easier.

The key difference is:
| `merge()`                                                   | `concat()`                             |
| ----------------------------------------------------------- | -------------------------------------- |
| Joins DataFrames using a common column (like `Customer_ID`) | Stacks or combines DataFrames directly |
| Similar to SQL JOIN                                         | Similar to appending tables together   |


📘 What is concat()?

concat() is used to combine DataFrames or Series.

There are two ways:

Vertical (axis=0) → Stack rows (default)
Horizontal (axis=1) → Join columns side by side

```python
import pandas as pd

df1 = pd.DataFrame({
    "Name": ["Rahul", "Neha"],
    "Salary": [50000, 60000]
})

df2 = pd.DataFrame({
    "Name": ["Aman", "Priya"],
    "Salary": [45000, 70000]
})

#Print both DataFrames.
print(df1)
print()
print(df2)

print()

#Concatenate them vertically.
result = pd.concat([df1, df2], ignore_index=True,axis=0)
print(result)

#How many rows are there after concatenation?
print(result.shape)

#Concatenate them using:ignore_index=True
result1 = pd.concat([df1, df2], ignore_index=True)
print(result1)

```

### Horizontal Concatenation (axis=1)

```python
import pandas as pd

names = pd.DataFrame({
    "Name": ["Rahul", "Neha", "Priya", "Aman"]
})

salary = pd.DataFrame({
    "Salary": [50000, 60000, 70000, 45000]
})

city = pd.DataFrame({
    "City": ["Pune", "Mumbai", "Delhi", "Hyderabad"]
})

#Print all three DataFrames.
print(names)
print(salary)
print(city)

#Concatenate names and salary horizontally.
res = pd.concat([names, salary],axis=1)
print(res)

#Concatenate all three DataFrames horizontally.
result = pd.concat([names, salary, city], axis=1)
print(result)

#How many columns are there after concatenating all three DataFrames?
print(result.shape)
```

💡 Rule to memorize forever
1. axis=0 → Add more records (rows).
2. axis=1 → Add more information (columns).

### 📘 concat() with Different Number of Rows

```python
import pandas as pd

names = pd.DataFrame({
    "Name": ["Rahul", "Neha", "Priya", "Aman"]
})

salary = pd.DataFrame({
    "Salary": [50000, 60000]
})

print(names)
print()
print(salary)

#Horizontal Concatenation
result = pd.concat([names, salary], axis=1)
print(result)

# 🤔 Why NaN?

# Because:

# Row 0 matched with Row 0 ✅
# Row 1 matched with Row 1 ✅
# There is no row 2 in salary
# There is no row 3 in salary

# So Pandas fills the missing values with NaN.
```

### Practice

```python
import pandas as pd

students = pd.DataFrame({
    "Name": ["Rahul", "Neha", "Priya", "Aman", "Karan"]
})

marks = pd.DataFrame({
    "Maths": [90, 85, 95]
})

#Print both DataFrames.
print(students)
print(marks)

#Concatenate them horizontally.
result = pd.concat([students, marks],axis=1)
print(result)

#How many NaN values are there in the output?
print(result.isna().sum().sum())
# NaN appears because the students DataFrame has 5 rows, while the marks DataFrame has only 3 rows. 
# During horizontal concatenation (axis=1),Pandas matches rows by their index.
# Since there are no marks for indexes 3 and 4, Pandas fills those missing values with NaN.

#What is the shape of the concatenated DataFrame?
print(result.shape)
```
