### Filtering DataFrames

🧠 What is Filtering?

Filtering means:

Selecting only the rows that satisfy a condition.

Example:

Suppose we have:

Name	Salary
Rahul	45000
Aditi	52000
Neha	68000
Vikram	90000

If we want employees earning more than ₹60,000:

df["Salary"] > 60000

Output:

0    False
1    False
2     True
3     True
Name: Salary, dtype: bool

This is called a Boolean Series.

Step 2

Now use it to filter.

df[df["Salary"] > 60000]

Output:

Name	Salary
Neha	68000
Vikram	90000

Notice the syntax:

df[ condition ]

This is one of the most important patterns in Pandas.

### Comparison Operators

| Operator | Meaning                  |
| -------- | ------------------------ |
| `>`      | Greater than             |
| `<`      | Less than                |
| `>=`     | Greater than or equal to |
| `<=`     | Less than or equal to    |
| `==`     | Equal to                 |
| `!=`     | Not equal to             |


```python
import pandas as pd

df = pd.read_csv('employee_data.csv')
#print(df)

print(df['Salary'] > 60000)  #this gives in boolean series
print(df[df['City'] == 'Pune'])  #this gives the value

#Multiple Conditions

#AND(&)
print(df[
    (df['Gender'] == "Female") &
    (df['Department'] == 'IT')
])

#OR(|)
print(df[
    (df["City"] == "Pune") |
    (df["City"] == "Mumbai")
])

#NOT(~)
print(df[
    ~(df["City"] == "Pune")
])
```

### Selecting Specific Columns After Filtering

```python
print(df.loc[
    df['Salary'] > 60000, 
    ['Name', 'Department','City']
    ])

print(df[
    (df['Department'] == 'IT') &
    (df['Salary'] > 80000) &
    (df['Experience'] > 5)
])
```

### 📝 Practice Questions

```python
import pandas as pd

df = pd.read_csv('employee_data.csv')
#print(df)

#Q1 Display employees whose salary is greater than ₹60,000.
print(df[df['Salary'] > 60000])

#Q2 Display employees from Pune.
print(df[df['City'] == 'Pune'])

#Q3 Display employees whose Age is less than 28.
print(df[df['Age'] < 28])

#Q4 Display employees whose Department is IT.
print(df[df['Department'] == 'IT'])

#Q5 Display employees whose Experience is greater than or equal to 5 years.
print(df[df['Experience'] >= 5])

#Q6 Display employees whose Salary is between ₹50,000 and ₹80,000.
print(df[
    (df['Salary'] >= 50000) &
    (df['Salary'] <= 80000)])

#print(df[df["Salary"].between(50000, 80000)])

#Q7 Display employees who are: from Pune AND work in IT
print(df[
    (df['City'] == 'Pune') &
    (df['Department'] == 'IT')
])

#Q8 Display employees who are: from Pune OR Mumbai
print(df[(df['City'] == 'Pune') |
    (df['City'] == 'Mumbai')])

#Q9 Display employees who are NOT from Hyderabad.
print(df[
    ~(df['City'] == 'Hyderabad')
])

#Q10 Display only the Name, Department, and Salary of employees who earn more than ₹70,000.
print(df.loc[
    df['Salary'] > 70000,
    ['Name','Department','Salary']
])

#Q11 Display only the Name and City of female employees working in HR.
print(df.loc[
    (df['Department'] == 'HR') &
    (df['Gender'] == 'Female'),
    ['Name','City']
])

#Q12
# The HR department wants a list of employees who satisfy all these conditions:

# Salary > ₹60,000
# Experience ≥ 5
# Performance Rating ≥ 4.5

# Display only these columns:

# Name
# Department
# Salary
# Experience
# Performance_Rating
print(df.loc[
    (df['Salary'] > 60000) &
    (df['Experience'] >= 5) &
    (df['Performance_Rating'] >= 4.5),
    ['Name','Department','Salary','Experience','Performance_Rating']
])
```
