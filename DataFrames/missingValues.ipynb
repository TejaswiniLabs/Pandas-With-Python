### Handling Missing Values (NaN)

🤔 What is a Missing Value?

Sometimes data is incomplete.

Example:

Name	Salary	City
Rahul	45000	Pune
Aditi	NaN 	Mumbai
Neha	68000	NaN
Aman	40000	Delhi

Here:

Aditi's Salary is missing.
Neha's City is missing.

Pandas represents missing values as:

NaN

which stands for Not a Number.

1️⃣ Creating Missing Values

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["Rahul", "Aditi", "Neha", "Aman"],
    "Salary": [45000, np.nan, 68000, 40000],
    "City": ["Pune", "Mumbai", np.nan, "Delhi"]
})

print(df)
```

```python
#2️⃣ Detect Missing Values
print(df.isna())  #gives in boolean values where there is NaN it will give you True otherwise false
print(df.isnull()) #It gives the same output as isna().

#3️⃣ Count Missing Values
print(df.isna().sum()) #This tells us how many missing values are present in each column.

#4️⃣ Total Missing Values
print(df.isna().sum().sum()) #There are 2 missing values in the whole DataFrame.

#5️⃣ Remove Missing Values
print(df.dropna()) #Rows with any missing value are removed.

#6️⃣ Fill Missing Values
df['Salary'] = df['Salary'].fillna(45000)
print(df)

df['City'] = df['City'].fillna('Banglore')
print(df)

#7️⃣ Fill Using Mean
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
print(df)

print(df.isna().sum())
```

### 📝 Practice Dataset

```python
import pandas as pd
import numpy as np

employee = {
    "Name": ["Rahul", "Aditi", "Neha", "Aman", "Priya"],
    "Salary": [45000, np.nan, 68000, 52000, np.nan],
    "City": ["Pune", "Mumbai", np.nan, "Delhi", "Hyderabad"],
    "Rating": [4.2, np.nan, 4.8, 4.1, 4.7]
}

df = pd.DataFrame(employee)

#Q1 Print the DataFrame.
print(df)

#Q2 Show which values are missing.
print(df.isna())

#Q3 Count missing values in every column.
print(df.isna().sum())

#Q4 Find the total number of missing values.
print(df.isna().sum().sum())

#Q5 Drop all rows containing missing values.
print(df.dropna())

#Q6 Replace missing Salary with 0.
df['Salary'] = df["Salary"].fillna(0)
print(df)

#Q7 Replace missing City with "Unknown".
df['City'] = df["City"].fillna('Unknown')
print(df)

#Q8 Replace missing Rating with the average rating.
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
print(df)

#Q9 Without modifying the original DataFrame:

# Create a new DataFrame called clean_df.
# Fill:
# Salary → average salary
# City → "Unknown"
# Rating → average rating

# (Hint: Use copy() so the original DataFrame stays unchanged.)
clean_df = df.copy()


clean_df["Salary"] = clean_df["Salary"].fillna(clean_df["Salary"].mean())

clean_df["City"] = clean_df["City"].fillna("Unknown")

clean_df["Rating"] = clean_df["Rating"].fillna(clean_df["Rating"].mean())

print(clean_df)
print(df.dropna().shape)
```
