### Selecting Rows and Columns Together

Until now you've learned:

1. df["Salary"]      # Column
2. df.loc[0]         # Row by label
3. df.iloc[0]        # Row by position

Now we'll combine them.

Syntax of .loc

df.loc[row_label, column_label]

Remember:

First comes the row
Then comes the column

Think of it as:

df.loc[ROW, COLUMN]

```python
import pandas as pd

df = pd.read_csv('employee_data.csv')
#print(df)

#.loc[]
print(df.loc[0:4])

#Rows + Columns
print(df.loc[0,'Name']) 
print(df.loc[0,['Name','Salary','Age']])
#Slicing
print(df.loc[0:10,['Name','Department','Age','Experience','Salary']])
```

### iloc[]

Syntax:

df.iloc[row_position, column_position]

Again:

ROW , COLUMN

But both are positions, not labels.

```python
# Row position = 0
# Column position = 1 (Name)
print(df.iloc[0,1]) 

print(df.iloc[0:10,1:5])  #here 10 is exclusive
#With .iloc, slicing behaves like normal Python.
```

| `.loc`                           | `.iloc`                          |
| -------------------------------- | -------------------------------- |
| Uses labels                      | Uses positions                   |
| End value is included in slicing | End value is excluded in slicing |
| `df.loc[0:4]` → 0,1,2,3,4        | `df.iloc[0:4]` → 0,1,2,3         |
