### Selecting Data from a DataFrame

```python
print(df['Salary'])  #This is series
print(df[['Name', 'Salary']])  #This is Dataframe
print(type(df[['Salary','Name']]))
print(df["Salary"].shape)
print(df[['Name', 'Salary']].shape)

# 🤔 Why is this a DataFrame?

# Because the double brackets tell Pandas:

# "I'm giving you a list of columns."

# Even though the list contains only one column:

# ["Salary"]

# it's still a list.

# Pandas returns a DataFrame.

# Single []   → One column → Series
# Double [[]] → One or more columns → DataFrame

#This is one of the most important rules in Pandas.

# df["Salary"] selects a single column, so Pandas returns a Series, which is one-dimensional. 
# df[["Salary"]] uses a list of column names, so Pandas treats the result as a DataFrame, which is two-dimensional, even if the list contains only one column.
```

### Accessing Rows (.loc and .iloc)

1. .loc[] → Access rows and columns by labels
2. .iloc[] → Access rows and columns by integer positions

```python
print(df.iloc[1])
print(type(df.iloc[1]))

print(df.loc[2])
print(type(df.loc[2]))
```

.loc uses labels, not positions.

But in your DataFrame, the labels happen to be 0, 1, 2, 3..., so they look exactly like positions.

Let's see why.

Case 1: Default Index

Your DataFrame currently has this index:

Index	Name
0	Rahul
1	Aditi
2	Neha
3	Aman

When you do:

df.loc[0]

Pandas thinks:

"Find the row whose label is 0."

The label is indeed 0, so it returns Rahul.

Now:

df.iloc[0]

Pandas thinks:

"Give me the row at position 0."

The first row is Rahul.

So both return the same row, but for different reasons.

Case 2: Change the Index

Now let's make Name the index.

df2 = df.set_index("Name")
print(df2)

Now the DataFrame becomes:

Name	Salary	City
Rahul	45000	Pune
Aditi	52000	Mumbai
Neha	68000	Delhi

Notice the index is no longer 0,1,2.

The labels are now:

Rahul
Aditi
Neha
...

Now look at this:

df2.loc["Neha"]

.loc searches by the label, which is "Neha".

✅ It returns Neha's row.

But what about:

df2.iloc[2]

.iloc doesn't care about the label.

It simply says:

"Give me the third row."

If Neha is the third row, you'll still get Neha.

Visual Understanding
.loc
Labels

Rahul
Aditi
Neha
Aman

You ask by name:

df2.loc["Neha"]
.iloc
Positions

0
1
2
3

You ask by position:

df2.iloc[2]

| `.loc`                                              | `.iloc`                                 |
| --------------------------------------------------- | --------------------------------------- |
| Uses **labels**                                     | Uses **positions**                      |
| `df.loc["Neha"]`                                    | `df.iloc[2]`                            |
| `df.loc[0]` works only because `0` is the row label | `df.iloc[0]` always means the first row |
