### Updating DataFrames (Add, Modify & Delete)

1️⃣ Add a New Column

```python
import pandas as pd

df = pd.read_csv('employee_data.csv')
#print(df)

#Suppose the company gives every employee a 10% salary hike.
df["New_Salary"] = df["Salary"] * 1.10
print(df)

#Add a Bonus Column
df["Total Salary"] = df['Salary'] + df['Bonus']
print(df)
```

2️⃣ Modify an Existing Column

```python
df['Salary'] = df['Salary'] + 2000
print(df)
```

3️⃣ Update Only Specific Rows

```python
df.loc[df["Name"] == "Rahul", "Salary"] = 50000
print(df)

df.loc[df['Department'] == 'IT', 'Salary'] += 7000
print(df)
```

4️⃣ Rename Columns

```python
df.rename(columns={'Performance_Rating':'Rating'}, inplace=True)
print(df)

df.rename(columns={"Employee_ID":'EmpId'}, inplace= True)
print(df)
```

5️⃣ Drop Columns and Rows

```python
# df.drop(index=0, inplace = True)
# print(df)

df.drop(index=[1, 2], inplace=True)
print(df)
```

7️⃣ Save the Updated DataFrame

```python
df.to_csv("updated_employee_data.csv", index=False)
print(df)
```

### 📝 Practice Questions

```python
import pandas as pd
df = pd.read_csv('employee_data.csv')
#print(df)

#Q1 Create a new column called: Annual_Salary Formula: Salary × 12
df['Annual_Salary'] = df['Salary'] * 12
print(df)

#Q2 Create a new column called: Total_Income Formula: Salary + Bonus
df['Total_Income'] = df['Salary'] + df['Bonus']
print(df)

#Q3 Increase everyone's salary by ₹3,000.
df['Salary'] = df['Salary'] + 3000
print(df)

#Q4 Increase the salary of employees in the IT department by ₹5,000.
df.loc[df['Department'] == 'IT', 'Salary'] += 5000
print(df) 

#Q5 Change Rahul's city to Bangalore.
df.loc[df['Name'] == "Rahul", 'City'] = 'Bangalore'
print(df)

#Q6 Rename: Performance_Rating to Rating
df.rename(columns={"Performance_Rating": "Rating"}, inplace = True)
print(df)

#Q7 Delete the Bonus column.
df.drop(columns='Bonus', inplace = True)
print(df)

#Q8 Delete the employee whose index is 3.
df.drop(index=3, inplace = True)
print(df)

#Q9 Create a new column: Tax
# Rule:
# Salary > ₹70,000 → Tax = 20% of Salary
# Otherwise → Tax = 10% of Salary

# Hint: You'll learn a cleaner way later with np.where(), but for now, try using .loc.

df.loc[df["Salary"] > 70000, "Tax"] = (
    df.loc[df["Salary"] > 70000, "Salary"] * 0.20
)
print(df)

#Q10 Create a new column: Status
# Rules:
# Rating ≥ 4.7 → "Excellent"
# Rating ≥ 4.3 and < 4.7 → "Good"
# Otherwise → "Average"

df["Status"] = "Average"
df.loc[df["Rating"] >= 4.3, "Status"] = "Good"
df.loc[df['Rating'] >= 4.7, 'Status'] = 'Excellent'
print(df)
```
