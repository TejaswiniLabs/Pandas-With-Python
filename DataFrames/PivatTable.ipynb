### 📘 Pivot Tables

A pivot table summarizes large amounts of data into an easy-to-read table. It's similar to the Pivot Table feature in Excel.

### Why do we use Pivot Tables?

Imagine this employee data:
| Name  | Department | City   | Salary |
| ----- | ---------- | ------ | -----: |
| Rahul | IT         | Pune   |  45000 |
| Neha  | HR         | Mumbai |  52000 |
| Aman  | IT         | Pune   |  60000 |
| Priya | HR         | Pune   |  55000 |
| Sneha | IT         | Mumbai |  70000 |

Suppose your manager asks:

"What is the average salary of each department?"

You already know how to do it:

df.groupby("Department")["Salary"].mean()

Now imagine they ask:

"Show the average salary of each department in each city."

This is where Pivot Tables become very useful.

Syntax
df.pivot_table(
    values="Salary",
    index="Department",
    aggfunc="mean"
)

| Parameter | Meaning                                      |
| --------- | -------------------------------------------- |
| `values`  | Which column to calculate                    |
| `index`   | Group by this column                         |
| `aggfunc` | Operation (mean, sum, max, min, count, etc.) |


```python
import pandas as pd

employee = {
    "Name": ["Rahul","Neha","Aman","Priya","Sneha"],
    "Department": ["IT","HR","IT","HR","IT"],
    "Salary": [45000,52000,60000,55000,70000]
}

df = pd.DataFrame(employee)

print(df.pivot_table(
    values = "Salary",
    index = "Department",
    aggfunc= "mean"
))

print(df.groupby("Department")["Salary"].mean())

# Both produce almost the same result.

# So why learn Pivot Tables?

# Because they can easily summarize multiple dimensions (rows, columns, totals) with much less code.
# Both groupby() and pivot_table() can produce similar results. However, pivot_table() is more flexible
# because it can summarize data using multiple rows, columns, and aggregation functions in one table,
# making it very useful for reporting and data analysis.
```

### 📝 Practice

```python
import pandas as pd

employee = {
    "Name": ["Rahul","Neha","Aman","Priya","Sneha","Rohan","Karan","Aditi"],
    "Department": ["IT","HR","IT","Finance","HR","Finance","IT","HR"],
    "City": ["Pune","Mumbai","Pune","Delhi","Mumbai","Delhi","Pune","Pune"],
    "Salary": [45000,52000,60000,70000,55000,65000,58000,50000]
}

df = pd.DataFrame(employee)

print(df)

#Q1 Find the average salary of each department using pivot_table().
print("Average Salary:\n",df.pivot_table(
    values = "Salary",
    index = "Department",
    aggfunc = "mean"
))

#Q2 Find the maximum salary of each department.
print("Maximum Salary:\n",df.pivot_table(
    values = "Salary",
    index = "Department",
    aggfunc = "max"
))

#Q3 Find the total salary of each department.
print("Total Salary:\n",df.pivot_table(
    values = "Salary",
    index = "Department",
    aggfunc = "sum"
))

#Q4 Find the number of employees in each department.
print("Number of Enployees:\n",df.pivot_table(
    values = "Name",
    index = "Department",
    aggfunc = "count"
))
```

### Multiple Indexes in Pivot Tables

```python
import pandas as pd

employee = {
    "Name": ["Rahul","Neha","Aman","Priya","Sneha","Rohan","Karan","Aditi"],
    "Department": ["IT","HR","IT","Finance","HR","Finance","IT","HR"],
    "City": ["Pune","Mumbai","Pune","Delhi","Mumbai","Delhi","Pune","Pune"],
    "Salary": [45000,52000,60000,70000,55000,65000,58000,50000]
}

df = pd.DataFrame(employee)

#📝 Practice
print(df.pivot_table(
    values = "Salary",
    index = ["Department","City"],     #This groups the data first by Department, then by City.
    aggfunc= ["mean","sum"]
))

#Q1 Find the average salary for each Department and City.
print(df.pivot_table(
    values = "Salary",
    index = ["Department","City"],     
    aggfunc= "mean"
))

#Q2 Find the maximum salary for each Department and City.
print(df.pivot_table(
    values = "Salary",
    index = ["Department","City"],
    aggfunc = "max"
))

#Q3 Find the total salary for each Department and City.
print(df.pivot_table(
    values = "Salary",
    index = ["Department","City"],
    aggfunc = "sum"
))

#Q4 Count the number of employees in each Department and City.
print(df.pivot_table(
    values = "Name",
    index = ["Department","City"],
    aggfunc = "count"
))
```

### 🚀 Now the Most Important Feature: columns=

```python
# The columns parameter creates separate columns for each unique value in a column (such as City).
#  This helps compare values side by side.
#  For example, we can compare the average salary of each department across different cities in one table.

import pandas as pd

employee = {
    "Name": ["Rahul","Neha","Aman","Priya","Sneha","Rohan","Karan","Aditi"],
    "Department": ["IT","HR","IT","Finance","HR","Finance","IT","HR"],
    "City": ["Pune","Mumbai","Pune","Delhi","Mumbai","Delhi","Pune","Pune"],
    "Salary": [45000,52000,60000,70000,55000,65000,58000,50000]
}

df = pd.DataFrame(employee)

print(df.pivot_table(
    values = "Salary",
    index = "Department",
    columns = "City",
    aggfunc = ["mean", "sum"]
))

# 🤔 What's happening?
# Rows = Department
# Columns = City
# Values = Salary
# Calculation = Mean

# This creates a table similar to an Excel Pivot Table.

# 📝 Practice
#Q1 Find the average salary for each Department, with City as columns.
print(df.pivot_table(
    values = "Salary",
    index = "Department",
    columns = "City",
    aggfunc = "mean"
))

#Q2 Find the maximum salary for each Department, with City as columns.
print(df.pivot_table(
    values = "Salary",
    index = "Department",
    columns = "City",
    aggfunc = "max"
))

#Q3 Find the total salary for each Department, with City as columns.
print(df.pivot_table(
    values = "Salary",
    index = "Department",
    columns = "City",
    aggfunc = "sum"
))

#Q4 Count the number of employees in each Department, with City as columns.
print(df.pivot_table(
    values = "Name",
    index = "Department",
    columns = "City",
    aggfunc = "count"
))
```
