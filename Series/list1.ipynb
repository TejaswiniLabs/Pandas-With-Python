```python
#A list is the most common Python data structure.
import pandas as pd

marks = [67, 45, 57, 89, 76]
print("Before making it into Series:",marks)
convert_marks = pd.Series(marks)
print(convert_marks)
```

### 🧠 Why doesn't Pandas just keep it as a list?

A Python list only stores values.
[85,90,78,92]

A Series stores extra information (metadata):

Index
Data type
Methods
Labels

This allows you to do things like:
series_marks.mean()
series_marks.max()
series_marks.min()
series_marks.sum()
A plain Python list doesn't have these built-in data analysis methods.

🧠 Important Concept

Many beginners think:

Series = List
This is not true.

Think of it like this:
Python List
     │
     ▼
Pandas Series
     │
     ├── Values
     ├── Index
     ├── Data Type
     └── Hundreds of useful methods
A Series is much more powerful than a list.


```python
ages = [20, 21, 22, 23, 24]
series_ages = pd.Series(ages)
print(series_ages)

cities = [
    "Pune",
    "Mumbai",
    "Delhi",
    "Hyderabad"
]
series_cities = pd.Series(cities)
print(series_cities)
print(type(series_cities))    #the type of series_cities is <class 'pandas.core.series.Series'> or <class 'pandas.Series'>
print(series_cities.dtype)

numbers = pd.Series([5, 10, 15])
print(numbers)
print(type(numbers))
print(numbers.dtype)

numbers = [5, 10, 15]
print(type(numbers))

```
