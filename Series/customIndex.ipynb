### What is Custom Index?

A custom index is a user-defined index assigned to a Pandas Series instead of the default integer index. It makes data more meaningful and allows label-based access to values.

Think about a real-life dataset (students, employees, products, cricket players, movies, etc.).

If you were creating a Series for that dataset:

What would you use as the index?
What would you store as the values?

For example:

Dataset: Cricket players

Index → Player names
Values → Runs scored

```python
#Creating Custom Index
import pandas as pd
students = pd.Series([11, 12, 13, 14], index = ['Bhakti', 'Om', 'Teju', 'Somesh'])
print(students)
print("Roll no. of Somesh:",students['Somesh'])

products = pd.Series(
    [65000, 800, 1800],
    index=[
        "Laptop",
        "Mouse",
        "Keyboard"
    ]
)

print(products)
print("Price of Mouse:", products['Mouse'])

salary = pd.Series(
    [35000, 40000, 50000, 60000],
    index = [
        "IT",
        "HR",
        "Finance",
        "Marketing"
    ]
)
print("Salary of IT sector:",salary["IT"])
sal = salary["IT"] = 45000
print("Updated Salary of IT sector:",sal)
print(salary)

#Pandas create a new index if the index is not present in the series
salary["Developer"] = 70000
print(salary)
```
