### Creating a Series from a Scalar Value

```python
#This lesson introduces a new concept called broadcasting.

import pandas as pd
s = pd.Series(100)
print(s)

s = pd.Series(
    100,
    index=["A", "B", "C", "D"]
)
print(s)

# Think of it like a teacher announcing:
# "Everyone gets 100 marks."

# If there are four students:
# | Student | Marks |
# | ------- | ----: |
# | A       |   100 |
# | B       |   100 |
# | C       |   100 |
# | D       |   100 |
# The same value is assigned to everyone.
#Broadcasting is the process of repeating a single value across multiple indexes or positions.

# This is called broadcasting.


temperature = pd.Series(
    25,
    index=["Monday", "Tuesday", "Wednesday", "Thursday"]
)

print(temperature)
print(type(temperature))
print(temperature.dtype)
```
