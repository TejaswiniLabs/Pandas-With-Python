### What is Tuple?

❓ Do you think Pandas keeps this as a tuple internally, or does it convert it into its own Series data structure?
Pandas converts the tuple into a Series because a Series is Pandas' own data structure. Internally, Pandas needs data in the Series format so it can provide features like indexes, labels, data types (dtype), and methods such as .sum(), .mean(), and .max(). A tuple by itself doesn't have these capabilities.

A tuple is a Python data structure, just like a list.

### List vs Tuple

| List                     | Tuple                                        |
| ------------------------ | -------------------------------------------- |
| Uses `[]`                | Uses `()`                                    |
| Mutable (can be changed) | Immutable (cannot be changed after creation) |
| More commonly used       | Used when data shouldn't change              |


```python
import pandas as pd
nums = (10, 20, 30, 40, 50)
nums_series = pd.Series(nums)
print(nums_series)
```

### 🧠 Why Does Pandas Accept Both?

Here's a key Python concept.

Both lists and tuples are iterables.

An iterable is any object that can be looped over.
Example:
numbers = (10, 20, 30)

for i in numbers:
    print(i)
Since Pandas can read each value one by one, it can build a Series from it.

```python
import pandas as pd

months = (
    "January",
    "February",
    "March",
    "April"
)

month_series = pd.Series(months)

print(month_series)
print(type(month_series))
print(month_series.dtype)
```
