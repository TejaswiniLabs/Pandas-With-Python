student = {
    "Rahul": 90,
    "Aditi": 95,
    "Neha": 88
}

When we convert this dictionary to a Series,

🤔 Do you think the dictionary keys ("Rahul", "Aditi", "Neha") will become:
The index of the Series?

Yes it will be the index of the series as :
Here:

"Rahul", "Aditi", "Neha" are keys.
90, 95, 88 are values.

A Series also has two parts:

Index
Values

So Pandas naturally maps them like this:

| Dictionary | Series |
| ---------- | ------ |
| Key        | Index  |
| Value      | Value  |



```python
import pandas as pd

cities = {
    "Pune" : 10,
    "Mumbai" : 20,
    "Solapur" : 30
}

cities_df = pd.Series(cities)
print(cities_df)

prices = {
    "Laptop": 65000,
    "Mouse": 800,
    "Keyboard": 1800
}
price_series = pd.Series(prices)
print(price_series["Laptop"])


fruits = {
    "Apple": 120,
    "Mango": 90,
    "Banana": 40,
    "Orange": 80
}

fruit_series = pd.Series(fruits)

print(fruit_series)
print(fruit_series["Mango"])

nums = {
    1: "One",
    2: "Two"
}
print(pd.Series(nums))
print(pd.Series(nums, index=[1, 2, 3])) 
print(nums[1])
```

### There are two ways to access values in Series

1. Position-based = s[0]
2. Label-based = s["Apple"]
  
You might think it will return 90, but here's something important:

In some situations, it may return the first value.
In other situations, it may not.
Because of this ambiguity, modern Pandas encourages using:
1. s.loc["Rahul"] → label-based access
2. s.iloc[0] → position-based access
