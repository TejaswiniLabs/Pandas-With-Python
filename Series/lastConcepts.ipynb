### Vectorized Operations

🤔 What is Vectorization?
Let's first solve a problem using a normal Python list.

Suppose every employee gets a ₹5,000 bonus.

Using a Python list
salary = [35000, 42000, 51000]

new_salary = []

for s in salary:
    new_salary.append(s + 5000)

print(new_salary)

Output:

[40000, 47000, 56000]

It works, but we had to write a loop.


Now the Pandas way
import pandas as pd

salary = pd.Series([35000, 42000, 51000])

print(salary + 5000)

Output:

0    40000
1    47000
2    56000
dtype: int64

No loop!

Pandas automatically adds 5000 to every element.

This is called vectorization.

```python
import pandas as pd

salary = pd.Series(
    [50000, 60000, 70000, 80000],
    index=["Rahul", "Aditi", "Neha", "Priya"]
)
print(salary)

# Vectorized operations apply the same operation to every element in a Series or DataFrame without writing explicit loops.
# They produce cleaner, more readable code and are much faster because Pandas uses optimized NumPy operations internally.

#Arithmetic operations on Series
print("Added bonus:\n",salary + 5000)
print(salary - 200)
print("Salary after tax deduction:\n",salary * 0.8)
print("Half Salary:\n",salary / 2)
print("Integer Half Salary:\n",salary // 2)
print("Salary squared:\n",salary ** 2)
```

### String Vectorization

```python
names = pd.Series([
    "rahul",
    "aditi",
    "neha"
])
print(names)

#String operations on Series
print("Upper:\n",names.str.upper())
print("Capitalized:\n",names.str.capitalize())
print("Title:\n",names.str.title())
print("Length:\n",names.str.len())
print("Contains 'a':\n",names.str.contains("a"))
print("Replace 'a' with 'A':\n",names.str.replace("a", "A"))
print("Split:\n",names.str.split("a"))
print("Strip:\n",names.str.strip())
print("Starts with 'r':\n",names.str.startswith("r"))
print("Ends with 'h':\n",names.str.endswith("h"))
print("Find 'a':\n",names.str.find("a"))
print("Findall 'a':\n",names.str.findall("a"))
print("Is alphanumeric:\n",names.str.isalnum())

sal = pd.Series(
    [50000, 60000, 70000, 80000],
    index=["Rahul", "Aditi", "Neha", "Priya"]
)
print(sal)

#Comparison operations on Series
print("Salary greater than 60000:\n",sal > 60000) #tgis gives a boolean series
print(sal[sal > 60000])  #this gives the actual values of the series where the condition is true
```

### Why is Vectorization Better?

Imagine a dataset with 1 million rows.

A Python loop works, but it's slower and requires more code.

Vectorized operations:

✅ Shorter code
✅ Easier to read
✅ Faster execution
✅ Optimized through NumPy

That's why you'll see them everywhere in Pandas projects.

### Boolean Filtering

What is Boolean Filtering??

Boolean filtering is a technique in Pandas where we use a Boolean condition (True/False) to select only the rows that satisfy a specific condition. It helps us quickly extract relevant data without writing loops.

Suppose you have:

import pandas as pd

salary = pd.Series(
    [35000, 42000, 51000, 60000],
    index=["Rahul", "Aditi", "Neha", "Priya"]
)

Imagine your manager asks:

"Show me only employees earning more than ₹50,000."

In plain Python, you'd probably write a loop.

In Pandas:

salary[salary > 50000]

Output:

Neha     51000
Priya    60000
dtype: int64

One line. No loop.

🧠 How does this work?

Step 1:

salary > 50000

creates a Boolean Series:

Rahul    False
Aditi    False
Neha      True
Priya     True
dtype: bool

Step 2:

salary[salary > 50000]

Pandas reads that Boolean Series and says:

Keep rows where the value is True.

| Employee | Salary | Keep? |
| -------- | -----: | :---: |
| Rahul    |  35000 |   ❌   |
| Aditi    |  42000 |   ❌   |
| Neha     |  51000 |   ✅   |
| Priya    |  60000 |   ✅   |

Result:

Neha     51000
Priya    60000

### More Examples

1. Greater than : 
salary[salary > 40000]

2. Less than :
salary[salary < 50000]

3. Equal to : 
salary[salary == 42000]

4. Not equal to :
salary[salary != 42000]

5. Greater than or equal to :
salary[salary >= 51000]

6. Less than or equal to :
salary[salary <= 42000]

Boolean filtering is better because it uses Pandas' vectorized operations, which are optimized through NumPy. It requires less code, is easier to read and maintain, and performs much faster than iterating through rows with a Python for loop, especially on large datasets.

```python
marks = pd.Series(
    [90, 85, 95, 70, 88],
    index=["Rahul", "Aditi", "Neha", "Aman", "Priya"]
)

print(marks[marks > 90])
print(marks[marks >= 90])
print(marks[marks < 85])
print(marks[marks == 88])
marks[marks != 95]
```

### Missing Values

This is one of the most important topics in data analysis because real-world datasets are rarely perfect.

Imagine this employee data:
| Name  |      Salary |
| ----- | ----------: |
| Rahul |       50000 |
| Aditi | *(missing)* |
| Neha  |       70000 |
| Priya | *(missing)* |

How do we:

detect missing values?
count them?
remove them?
replace them with meaningful values?

Notice:

Some salaries are missing.

Pandas represents missing values as:

NaN

NaN means:

Not a Number

Even though the name says "Number", Pandas uses NaN to represent missing data in many situations.


```python
import pandas as pd
import numpy as np

marks = pd.Series(
    [90, np.nan, 80, np.nan, 95],
    index=["Rahul", "Aditi", "Neha", "Priya", "Aman"]
)

print(marks)

print("\nMissing values:")
print(marks.isna())

print("\nDetecting not NaN:")
print(salary.notna())

print("\nNumber of missing values:")
print(marks.isna().sum())

print("\nWithout missing values:")
print(marks.dropna())

print("\nFilled with 85:")
print(marks.fillna(85))

print("\nFilled with mean:")
print(marks.fillna(marks.mean()))

# 🤔 Why did int64 become float64?
# Originally:
# [50000, 70000]
# could be stored as: int64

# But NaN is a floating-point value internally.

# So Pandas converts the entire Series to:

# float64

# This is normal behavior.

# True → Missing value
# False → Value exists
```

Why Do We Fill Missing Values?

Imagine this dataset:

Student	Marks
Rahul	90
Aditi	NaN
Neha	95

If we remove Aditi:

We lose information.

Sometimes that's okay.

Sometimes it's better to estimate the missing value.

Choosing between dropping and filling depends on the problem you're solving.

Use .dropna() when the missing rows are few or not important, and removing them won't significantly affect your analysis.
Use .fillna() when you want to keep all rows because the missing data is valuable or removing it would result in losing too much information.
