### Slicing a Series (Slicing by Position)

```python
import pandas as pd

marks = pd.Series(
    [90, 85, 95, 88, 76],
    index=["Rahul", "Aditi", "Neha", "Priya", "Aman"]
)

print(marks)

#Slicing by positional index
print("Slicing by positional index:",marks[1:4]) 

#Slicing from Beginning to a specific index
print("Slicing from Beginning to a specific index:",marks[:3])

#Slcing from a specific index to the end
print("Slicing from a specific index to the end:",marks[2:])

#Entire series slicing
print("Slicing entire series:",marks[:])
```

###  Label-based Slicing (Ending label is INCLUDED)

```python
temperature = pd.Series(
    [30, 31, 29, 32, 28, 27, 33],
    index=[
        "Mon",
        "Tue",
        "Wed",
        "Thu",
        "Fri",
        "Sat",
        "Sun"
    ]
)
print(temperature)
print(temperature['Mon':'Fri'])
print(temperature.loc['Thu':'Sun'])  #using loc to slice from Thu to Sun
print(temperature['Mon':])
print(temperature[:'Fri'])
print(temperature[:])
```

Why does s[1:4] exclude 4 but s["B":"D"] includes "D"?

The answer is:

Position slicing follows Python's slicing rules

Python has always worked like this:

numbers = [10,20,30,40,50]

numbers[1:4]

Output:

20
30
40

The ending index (4) is excluded.

Pandas follows this same rule when you're slicing by position.


But label slicing is different.

Imagine you have a Series:

| Label | Sales |
| ----- | ----: |
| Jan   |   100 |
| Feb   |   200 |
| Mar   |   300 |
| Apr   |   400 |
| May   |   500 |

Now imagine your manager says:

"Show me the sales from January to March."

If Pandas excluded "Mar" like Python excludes the stop index, the result would be:

Jan
Feb

That would feel surprising because when people say "January to March", they usually expect March to be included.

So the Pandas developers made a design decision:

When slicing by labels, include both the start and the end labels.

This makes label slicing more natural for working with real-world data.

Because label-based slicing is designed to be intuitive for real-world data analysis. When users specify a range of labels, they generally expect both the starting and ending labels to be included. Position-based slicing, on the other hand, follows Python's standard slicing rules where the stop position is excluded.

### Series Attributes
### 🤔 What is an Attribute?

Imagine you have a person.

Person
│
├── Name
├── Age
├── Height
└── Weight

These are attributes (properties) of the person.

Similarly, a Pandas Series also has attributes that tell us about the Series.

Think of them as the identity card of a Series.

```python
marks = pd.Series(
    [90, 85, 95, 88],
    index=["Rahul", "Aditi", "Neha", "Priya"]
)

print(marks)
print(marks.index)  #gives the index of the series
print(marks.values) #gives the values of the series in the form of an numpy array
print(marks.items()) #gives the index and values of the series in the form of a tuple
print(marks.dtype)
print(marks.shape) #gives the shape of the series  Python treats it as an integer, not a tuple. That's why there is comma after the shape value
print(marks.size) #gives the size of the series

marks.name = "Marks of Students"
print(marks.name) #gives the name of the series

```

| Attribute | Purpose              | Example Output              |
| --------- | -------------------- | --------------------------- |
| `.index`  | Returns index labels | `Index(['Rahul', 'Aditi'])` |
| `.values` | Returns data values  | `array([90,85])`            |
| `.dtype`  | Data type of values  | `int64`                     |
| `.shape`  | Shape of the Series  | `(4,)`                      |
| `.size`   | Number of values     | `4`                         |
| `.name`   | Name of the Series   | `"Student Marks"`           |


What is the difference between .shape and .size?

1. .size
Suppose we have:

salary = pd.Series(
    [35000, 42000, 51000]
)
salary.size

Output:

3

It tells you:

How many values are in the Series?

2. .shape
salary.shape

Output:

(3,)

It tells you:

What is the structure (dimensions) of the Series?

For a Series, that structure is one-dimensional, so the shape is a tuple with one value.
