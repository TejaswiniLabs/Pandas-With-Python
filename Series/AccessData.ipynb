### Accessing by Label

🏢 Real-Life Analogy

Imagine a phone contacts app.

Instead of saying:

"Give me contact number 53."

You say:

"Give me Rahul's contact."

Much easier, right?

That's why label-based indexing is powerful.

```python
import pandas as pd

marks = pd.Series(
     [90, 85, 95, 88],
     index=["Rahul", "Aditi", "Neha", "Priya"]
)
print(marks)
print("Accessing marks of Neha:",marks["Neha"])
marks["Somesh"] = 92
print("Added another student marks:",marks["Somesh"])
print(marks)
```

### Accessing Multiple Labels

```python
import pandas as pd

salary = pd.Series(
    [80000, 75000, 85000, 90000],
    index=['Om', 'Vaishnavi', 'Rohit', 'Raj']
)
print(salary)

# The outer [] is used to access data from the Series.
# The inner [] is a Python list containing the labels.
print("Salary of Om and Raj:\n", salary[['Om','Raj']])
```

### Accessing By Position(iloc, loc)

```python
#Sometimes you don't know the label. You only know the position.
import pandas as pd

contact = pd.Series(
    [9866754632, 9876543233, 9876543215, 9876543210],
    index = ['Vaishnavi', 'Mayur', 'Mahesh', 'Varsha']
)
print(contact)
print("Contact of Mayur:", contact.iloc[1])  #iloc is used to access data by position.
print("Contact of Varsha:", contact.loc["Varsha"]) #loc is used to access data by label.

#But later, when we work with DataFrames, .loc becomes much more powerful.

# | `.loc`               | `.iloc`                         |
# | -------------------- | ------------------------------- |
# | Label-based indexing | Integer position-based indexing |
# | Uses index labels    | Uses positions (0, 1, 2...)     |
```

| Method            | Example                    | Uses                   |
| ----------------- | -------------------------- | ---------------------- |
| Label             | `marks["Rahul"]`           | Access by label        |
| Multiple Labels   | `marks[["Rahul", "Neha"]]` | Access multiple labels |
| Position          | `marks.iloc[0]`            | Access by position     |
| Label with `.loc` | `marks.loc["Rahul"]`       | Explicit label access  |
