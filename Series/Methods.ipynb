### Series Methods

```python
import pandas as pd

salary = pd.Series(
    [30000, 45000, 25000, 60000, 60000, 40000],
    index = ['Rahul', 'Aditi', 'Neha', 'Priya', 'Amit', 'Sneha']
)
print(salary)

#Sum of all salaries
print("Total Salary of all employees:",salary.sum())

#Average of all salaries
print("Avg of all salary:",salary.mean())

#Highest salary among all
print("Highest salary among all:",salary.max())

#Lowest salary among all
print("Lowest salary:",salary.min())

#Count of all salaries
print("Count od all salary:",salary.count())

#Describe the salary series
print("Describe:\n",salary.describe())  #Think of .describe() as a summary report.

#Median of all salaries
print("Median:",salary.median())

#Sort the salary series by index
print("Sort by index:\n",salary.sort_index())

#Sort the salary series by values
print("Sort by values:\n",salary.sort_values())
print("Sort by values in descending order:\n",salary.sort_values(ascending=False))

#Unique 
print("Unique values:", salary.unique())

#Count of unique values
print("Count of unique values:", salary.nunique())

#Vlaue counts of unique values
print("Value counts:\n", salary.value_counts())
```

| Method           | What it returns                        |
| ---------------- | -------------------------------------- |
| `count()`        | Total number of **non-missing values** |
| `unique()`       | Only the **distinct values**           |
| `value_counts()` | **Frequency** of each unique value     |


### 📊 Summary

| Method            | Purpose                  |
| ----------------- | ------------------------ |
| `.sum()`          | Total                    |
| `.mean()`         | Average                  |
| `.max()`          | Highest value            |
| `.min()`          | Lowest value             |
| `.count()`        | Count non-missing values |
| `.describe()`     | Statistical summary      |
| `.sort_values()`  | Sort by values           |
| `.sort_index()`   | Sort by index            |
| `.unique()`       | Unique values            |
| `.value_counts()` | Frequency of values      |
