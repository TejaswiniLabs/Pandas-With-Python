```python
import pandas as pd
df = pd.read_csv('student_data.csv')
#print(df)

#Q1 How many students are there?
print(df.shape[0])

#Q2 How many columns are there?
print(df.columns.size)

#Q3 What are the names of all the columns?
print(df.columns)

#Q4 What is the data type of each column?
print(df.dtypes)

#Q5 Display only the Maths column.
print(df['Maths'])

#Q6 Display only the Name column.
print(df['Name'])

#Q7 What is the type of: df["Maths"] Is it a Series or a DataFrame? Explain why.
# A DataFrame is made up of multiple columns.

# When you select one column:

# df["Maths"]

# Pandas returns a Series.

# Think of it like this:
# DataFrame

# +-------+-------+---------+
# | Name  | Maths | Science |
# +-------+-------+---------+
# | Rahul | 88    | 91      |
# | Neha  | 95    | 93      |
# +-------+-------+---------+

#         ↓

# df["Maths"]

# 88
# 95

# ↓

# Series

# Selecting a single column from a DataFrame returns a Series because each column in a DataFrame is represented internally as a Pandas Series.

#Q8 Display the first 5 Maths marks.
print(df.Maths.head(5))

#Q9 Display the last 5 student names.
print(df.Name.tail(5))
print(df['Name'].tail(7))

#Q10
maths = df['Maths']
print(type(maths))
```

```python
# Q1
# Find:
# Highest Maths marks
# Lowest Maths marks
# Average Maths marks
print("Highest marks of maths:",df['Maths'].max()) 
print("Lowest marks of maths:",df['Maths'].min())
print("Average marks of maths:",df['Maths'].mean())


# Q2
# Find:
# Highest Attendance
# Lowest Attendance
# Average Attendance
print("Highest Attendance:",df['Attendance'].max())
print("Lowest Attendance:",df['Attendance'].min())
print("Average Attendance:",df['Attendance'].mean())

# Q3
# How many students belong to:
# Pune
# Mumbai
# Delhi
# Hyderabad
print("Number of students from Pune:",df[df['City'] == 'Pune'].shape[0])  #shape[0] gives the number of rows in the filtered DataFrame
print("Number of students from Mumbai:",df[df['City'] == 'Mumbai'].shape[0])
print("Number of students from Delhi:",df[df['City'] == 'Delhi'].shape[0])
print("Number of students from Hyderabad:",df[df['City'] == 'Hyderabad'].shape[0])

#Q4 Find the unique departments.
print(df['Department'].unique()) 

#Q5 How many students are in each department?
print(df['Department'].value_counts())

#Q6 Sort the Maths marks in ascending order.
print(df['Maths'].sort_values())

#Q7 Sort the Maths marks in descending order.
print(df['Maths'].sort_values(ascending=False))  #descending order

#Q8 How many students scored more than 90 in Maths?
print("Number of students who scored more than 90 in Maths:", df[df['Maths'] > 90].shape[0])

#Q9 Display the names of students who scored more than 90 in Maths.
print(df[df['Maths'] > 90].Name) #this gives the names of the students who scored more than 90 in Maths

#Q10 Which column in this dataset do you think would be most useful for predicting whether a student gets a scholarship?
#The subjects marks (Maths, Science, etc.) and Attendance would likely be the most useful columns for predicting whether
#a student gets a scholarship, as they directly reflect the student's academic performance and engagement.
```

```python
# Challenge 1
# Find the student who scored the highest Maths marks.Not just the mark.The entire student record.
print("Highest marks in maths subject:\n",df[df['Maths'] == 98])

# Challenge 2 
# Find the student with the highest attendance.
highest_attendance = df['Attendance'].max()
print("Highest Attendance:\n",df[df['Attendance'] == highest_attendance])

# Challenge 3
# Show all students who: scored above 90 in Maths and have attendance above 95
print(
    df[
        (df["Maths"] > 90) &
        (df["Attendance"] > 95)
    ]
)

# Challenge 4
# Show students who are from Pune and are in the MCA department
print(
    df[
        (df['City'] == 'Pune') &
        (df['Department'] == 'MCA')
    ]
)

# Challenge 5 
# Create a new column: Total which is: Maths + Science + English + Computer. Then find the top 5 students based on the total marks.
df["Total"] = (
    df["Maths"]
    + df["Science"]
    + df["English"]
    + df["Computer"]
)
print(df["Total"])
print(
    df.sort_values(
        by="Total",
        ascending=False
    )
)
```
