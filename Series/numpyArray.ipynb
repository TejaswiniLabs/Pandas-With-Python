### What is a NumPy array?

import numpy as np

arr = np.array([10, 20, 30, 40])

This is a NumPy array.

Now the question is:

Can Pandas convert this NumPy array into a Series?

✅ Yes!

### 💡 Why can Pandas do this?

Remember from Lesson 1:

Pandas is built on top of NumPy.

That means Pandas uses NumPy internally for storing and processing data.

Think of it like this:

Python
   │
   ▼
NumPy
   │
   ▼
Pandas

Pandas doesn't replace NumPy—it extends it with extra features like indexes, labels, and data analysis tools.
NumPy is mainly designed for fast numerical computations on arrays. Pandas is designed for real-world data analysis. It supports labels, indexes, text, dates, missing values, mixed data types, filtering, grouping, merging, and many other operations that are essential for working with structured datasets.

```python
import numpy as np
import pandas as pd

arr = np.array([10, 20, 30, 40])
print(arr)
#avg = np.mean(arr)
print("Avg using numpy:",arr.mean())
arr_series = pd.Series(arr)
print(arr_series)
print("Avg using pandas:",arr_series.mean())

marks = np.array([75, 80, 90, 95])
print("Marks array:", marks[0])
marks_series = pd.Series(marks, index=["Aditi", "Rohit", "Sita", "Ram"])

print(marks_series)
print(type(marks_series))
print(type(marks))
print(marks_series.dtype)



salary = pd.Series(
    [35000,42000,60000],
    index=["Rahul","Aditi","Neha"]
)

print(salary.mean())
print(salary["Neha"])
```

arr = np.array([10,20,30,40])

arr.mean()

NumPy thinks:

"I have numbers."

In Pandas:
arr_series = pd.Series(arr)

arr_series.mean()

Pandas thinks:

"I have numbers with labels."

It also calculates the average.

But it remembers:
| Index | Value |
| ----: | ----: |
|     0 |    10 |
|     1 |    20 |
|     2 |    30 |
