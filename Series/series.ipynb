### What is Series?

A Series is a one-dimensional labeled array.

That definition sounds technical, so let's break it down.

Imagine you have student marks:
85
90
78
92

This is just a list of values.

Now Pandas stores it like this:
| Index | Value |
| ----: | ----: |
|     0 |    85 |
|     1 |    90 |
|     2 |    78 |
|     3 |    92 |
This entire column is called a Series.

Think of it this way:
A Series = one column
A DataFrame = many columns

For example:
| Name  | Age | Marks |
| ----- | --: | ----: |
| Rahul |  20 |    85 |
| Aditi |  21 |    90 |
| Neha  |  20 |    78 |
Each column (Name, Age, Marks) is a separate Series.

👉 A DataFrame is made up of multiple Series placed side by side.

### What does "one-dimensional" mean?

One-dimensional means there is only one sequence of values.

For example:
10
20
30
40
There are no rows and columns like a table—just a single column of data.

### What does "labeled" mean?

Every value has a label, called an index.
Example:
| Index | Value |
| ----: | ----: |
|     0 |   100 |
|     1 |   200 |
|     2 |   300 |
Here:

0, 1, 2 are the labels (indexes).
100, 200, 300 are the values.

The index helps you locate data quickly.

Real-Life Analogy:
| Position | Contact |
| -------: | ------- |
|        0 | Rahul   |
|        1 | Aditi   |
|        2 | Neha    |
The position acts like the index, and the contact name is the value.

That's exactly how a Series works.

📌 Important Rule

A Series always has two parts:

Index
Values

If you don't specify an index, Pandas creates one starting from 0.

### Creating Your First Series

```python
import pandas as pd

marks = pd.Series([67, 89, 76, 45])
print(marks)

fruits = pd.Series(["Apple", "Banana", "Cherry", "Date"])
print(fruits)
```

Why do you think Pandas allows us to create a Series from so many different data types (list, tuple, dictionary, NumPy array, etc.) instead of only from a list?
In real-world applications, data comes from many different sources. It may already exist as a Python list, a tuple, a dictionary, a NumPy array, or even a database or CSV file. Pandas allows us to create a Series from different data types so we can easily analyze and manipulate the data regardless of its original format.
