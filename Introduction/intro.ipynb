### What are Pandas?

Imagine you have student data:

Merge multiple datasets.
Prepare data for machine learning.

Think of it like Excel, but programmable with Python.

### 🤔 Why Not Just Use NumPy?

| NumPy                              | Pandas                              |
| ---------------------------------- | ----------------------------------- |
| Works with arrays                  | Works with tables                   |
| Mostly numeric data                | Numeric + text + dates + mixed data |
| Faster for mathematical operations | Easier for data analysis            |
| No column names                    | Has row and column labels           |
| Best for scientific computing      | Best for real-world datasets        |






| NumPy                             | Pandas                                      |
| --------------------------------- | ------------------------------------------- |
| Arrays                            | Tables (DataFrames)                         |
| Mostly numerical data             | Mixed data (numbers, text, dates, booleans) |
| No column names                   | Has column names                            |
| Very fast mathematical operations | Excellent for data analysis                 |
| Foundation library                | Built on top of NumPy                       |


### The Two Main Data Structures

Everything in Pandas revolves around just two objects.

1️⃣ Series

A Series is a one-dimensional labeled array.

Think of it as a single column in Excel.

Example:
| Index | Value |
| ----: | ----: |
|     0 |    10 |
|     1 |    20 |
|     2 |    30 |
|     3 |    40 |


2️⃣ DataFrame

A DataFrame is a two-dimensional table.

Think of it as an entire Excel sheet.

Example:
| Roll | Name  | Marks |
| ---: | ----- | ----: |
|    1 | Rahul |    90 |
|    2 | Aditi |    95 |
|    3 | Neha  |    88 |

This is the most commonly used Pandas object.

### Why do we need Pandas?

Imagine a CSV file with 1 million rows.

Without Pandas:

Reading the file is difficult.
Searching or filtering is slow and requires lots of code.
Calculating summaries is tedious.

With Pandas, it's just a few lines of code.

import pandas as pd

df = pd.read_csv("students.csv")

print(df.head())

That's all it takes to load and preview the data
