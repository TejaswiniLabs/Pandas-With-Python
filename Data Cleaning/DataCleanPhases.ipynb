### 🟢 Phase 1 – Inspect the Data

```python
import pandas as pd

data = pd.read_csv("messy_orders.csv")
df = pd.DataFrame(data)
print(df)

print(df.head())

print(df.shape)

print(df.info())

print(df.describe())

print(df.isna())

print(df.isna().sum())

print(df.duplicated())

print(df.duplicated().sum())

print(df[df.duplicated()])
```

### 🚀 Phase 2: Start Cleaning the Data

```python
df["Customer"] = df["Customer"].str.strip()
print(df)

print(df["Customer"].str.strip())

df["Email"] = df["Email"].str.lower()
print(df)

print(df["Email"].isna().sum())

df["Email"] = df["Email"].fillna("unknown@gmail.com")
print(df)

df["Order_Date"] = pd.to_datetime(df["Order_Date"],format="mixed")
print(df)
```

### 🏢Phase 3: Data Cleaning & Feature Engineering

```python
import pandas as pd
import numpy as np

print(df["Quantity"].isna().sum())

df["Quantity"] = df["Quantity"].fillna(df["Quantity"].median())
print(df)

print(df["Discount"].isna().sum())

df["Discount"] = df["Discount"].fillna(0)
print(df)

print(df["Rating"].isna().sum())

df["Rating"] = df["Rating"].fillna(df["Rating"].mean())
print(df)

print(df.isna().sum())

print(df.duplicated().sum())

print(df.drop_duplicates(inplace=True))

print(df.shape)

df["Total_Amount"] = df["Price"] * df["Quantity"]
print(df)

df["Discount_Amount"] = df["Total_Amount"] * df["Discount"] / 100
print(df)

df["Final_Amount"] = df["Total_Amount"] - df["Discount_Amount"]
print(df)

df["Order_Month"] = df["Order_Date"].dt.month_name()
print(df)

df["Order_Year"] = df["Order_Date"].dt.year
print(df)

#Which product generated the highest total revenue?
product = df.groupby("Product")["Final_Amount"].sum()
print(product)
print(product.idxmax())

city = df.groupby("City")["Final_Amount"].sum()
print(city.sort_values(ascending=False))

payment = df.groupby("Payment_Method").size()
print(payment)

rating = df.groupby("Category")["Rating"].mean()
print(rating)
print(rating.idxmax())
```
