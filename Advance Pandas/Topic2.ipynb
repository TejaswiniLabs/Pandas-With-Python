### Next Topic: melt()
melt() converts wide data → long data.

```python
import pandas as pd

sales = {
    "Product": ["Laptop", "Mobile", "TV"],
    "Jan": [50000, 30000, 45000],
    "Feb": [55000, 35000, 47000],
    "Mar": [60000, 32000, 50000]
}

df = pd.DataFrame(sales)
print(df)

#Q1
result = df.melt(
    id_vars= "Product",
    var_name = "Month",
    value_name = "Sales"
)
print(result)

#Q2 How many rows are there after melting?
print(len(result))

#Q3 How many columns are there after melting?
print(result.shape[1])

#melt() converts wide data into long data by turning columns into rows.
```

| Function        | Purpose                                          |
| --------------- | ------------------------------------------------ |
| `melt()`        | Wide → Long                                      |
| `pivot()`       | Long → Wide (no duplicates allowed)              |
| `pivot_table()` | Long → Wide (duplicates allowed, uses `aggfunc`) |
