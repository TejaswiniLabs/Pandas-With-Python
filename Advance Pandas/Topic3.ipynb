### stack() and unstack()
1. stack() → Columns ➜ Rows
2. unstack() → Rows ➜ Columns

What is stack()?

stack() converts columns into rows.

```python
import pandas as pd

sales = {
    "Jan": [50000, 30000],
    "Feb": [55000, 35000]
}

df = pd.DataFrame(sales, index=["Laptop", "Mobile"])
print(df)

print(df.stack())  #coverted the column into row 
```

### What is unstack()?

It does the opposite. Row to Column

```python
import pandas as pd

sales = {
    "Jan": [50000, 30000],
    "Feb": [55000, 35000]
}

df = pd.DataFrame(sales, index=["Laptop", "Mobile"])
print(df)

result = df.stack()  
print(result.unstack())  #converted row to column to original one 
```
