### What is transform()?

Think of it like this:

1. groupby().mean() → Gives one row per group.
2. groupby().transform("mean") → Gives the group's mean for every original row.

```python
#Without transform()
import pandas as pd

employee = {
    "Name": ["Rahul","Neha","Aman","Priya","Karan","Sneha"],
    "Department": ["IT","HR","IT","HR","IT","HR"],
    "Salary": [50000,60000,70000,65000,55000,70000]
}

df = pd.DataFrame(employee)

print(df)
print(df.groupby("Department")["Salary"].mean())
#Only 2 rows because there are 2 departments.


#With transform
df["Dept_Avg"] = df.groupby("Department")["Salary"].transform("mean")
print(df)

# Notice:

# IT average is repeated for every IT employee.
# HR average is repeated for every HR employee.

# That's the whole purpose of transform().

#Is Rahul earning above or below his department's average?
df["Dept_Avg"] = df.groupby("Department")["Salary"].transform("mean")
df["Difference"] = df["Salary"] - df["Dept_Avg"]
print(df)
#Now you can immediately compare every employee with their department average.
```

```python
import pandas as pd

employee = {
    "Name": ["Rahul","Neha","Aman","Priya","Karan","Sneha","Rohan","Aditi"],
    "Department": ["IT","HR","IT","Finance","HR","Finance","IT","HR"],
    "Salary": [50000,60000,70000,80000,55000,75000,65000,50000]
}

df = pd.DataFrame(employee)

print(df)

#Q1
df["Dept_Avg_Salary"] = df.groupby("Department")["Salary"].transform("mean")
print(df)

#Q2
df["Difference"] = df["Salary"] - df["Dept_Avg_Salary"]
print(df)

#Q3 Display only employees whose salary is greater than their department average.
print(df[df["Salary"] > df["Dept_Avg_Salary"]])

#Q4 How many employees earn more than their department average?
print(df[df["Salary"] > df["Dept_Avg_Salary"]].shape[0])

#Q5 Difference between groupby().mean() and groupby().transform("mean")
# groupby().mean() returns one row for each group. 
# groupby().transform("mean") returns the group's mean for every row in the original DataFrame, so the number of rows stays the same.
```
