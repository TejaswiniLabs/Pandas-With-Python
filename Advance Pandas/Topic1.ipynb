### Topic 1: pd.crosstab()

What is crosstab()?

crosstab() counts how many times combinations of categories occur.

Think of it as a frequency table.

```python
import pandas as pd

employee = {
    "Department": ["IT","IT","HR","HR","Finance","Finance","IT"],
    "City": ["Pune","Mumbai","Pune","Delhi","Delhi","Pune","Pune"]
}

df = pd.DataFrame(employee)
print(df)
#crosstab
print(pd.crosstab(index= df["Department"], columns= df["City"]))
```

How to read it:
1. IT employees in Pune → 2
2. IT employees in Mumbai → 1
3. HR employees in Delhi → 1
4. Finance employees in Pune → 1

```python
import pandas as pd

employee = {
    "Name": [
        "Rahul","Neha","Aman","Priya",
        "Karan","Sneha","Rohan","Aditi",
        "Vikas","Meena"
    ],
    "Department":[
        "IT","HR","IT","Finance",
        "HR","Finance","IT","HR",
        "Finance","IT"
    ],
    "City":[
        "Pune","Mumbai","Pune","Delhi",
        "Pune","Delhi","Mumbai","Pune",
        "Delhi","Mumbai"
    ],
    "Gender":[
        "Male","Female","Male","Female",
        "Male","Female","Male","Female",
        "Male","Female"
    ]
}

df = pd.DataFrame(employee)

print(df)

#Practice Questions

#Q1 Create a crosstab of: Department vs City
print(pd.crosstab(index=df["Department"], columns= df["City"]))

#Q2 Create a crosstab of: Department vs Gender
print(pd.crosstab(index= df["Department"], columns= df["Gender"]))

#Q3 Create a crosstab of: City vs Gender
print(pd.crosstab(index= df["City"], columns= df["Gender"]))

#Q4 How many Male employees work in IT?
table = pd.crosstab(df["Department"], df["Gender"])
print(table.loc["IT", "Male"])

#Q5 How many Female employees are from Pune?
table1 = pd.crosstab(df["City"], df["Gender"])
print(table1.loc["Pune", "Female"])
```

### Topic: pivot() vs pivot_table()

```python
import pandas as pd

employee = {
    "Name": ["Rahul", "Neha", "Aman", "Priya"],
    "Department": ["IT", "HR", "Finance", "IT"],
    "Salary": [50000, 60000, 70000, 55000]
}

df = pd.DataFrame(employee)
print(df)

print(df.pivot(
    index = "Name",    #for rows
    columns = "Department",   #for columns
    values = "Salary"
))
```

| `pivot()`                                                    | `pivot_table()`                                                     |
| ------------------------------------------------------------ | ------------------------------------------------------------------- |
| Does **not** perform calculations.                           | Performs calculations using `aggfunc` (`mean`, `sum`, `max`, etc.). |
| Requires each `index + column` combination to be **unique**. | Can handle duplicate combinations because it aggregates them.       |
| Uses only `index`, `columns`, and `values`.                  | Uses `index`, `columns`, `values`, and usually `aggfunc`.           |


Interview Rule (Very Important)
1. ✅ Use pivot() when every index + column combination is unique.
2. ✅ Use pivot_table() when duplicate combinations exist or when you need calculations like mean, sum, count, etc.
