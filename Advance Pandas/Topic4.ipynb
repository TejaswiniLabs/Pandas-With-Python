## cut() & qcut()

### What is pd.cut()?

cut() divides data into fixed ranges (bins).

Example :
| Salary |
| -----: |
|  25000 |
|  45000 |
|  65000 |
|  85000 |

We want:

#### 0–40,000 → Low
#### 40,001–70,000 → Medium
#### Above 70,000 → High

```python
import pandas as pd

employee = {
    "Name": ["Rahul", "Neha", "Aman", "Priya", "Karan"],
    "Salary": [25000, 45000, 65000, 85000, 55000]
}

df = pd.DataFrame(employee)

df["Category"] = pd.cut(
    df["Salary"],
    bins=[0, 40000, 70000, 100000],
    labels=["Low", "Medium", "High"]
)

print(df)
```

### What are bins?

bins=[0, 40000, 70000, 100000]

creates these intervals:

0 --------40000--------70000--------100000

    Low       Medium        High

There are 4 numbers, so there are 3 categories.

Rule: Number of labels = Number of bins - 1

### What are labels?

labels=["Low", "Medium", "High"]

These are the names assigned to each interval.

```python
import pandas as pd

employee = {
    "Name": [
        "Rahul",
        "Neha",
        "Aman",
        "Priya",
        "Karan",
        "Sneha",
        "Rohan",
        "Aditi"
    ],
    "Salary": [
        28000,
        42000,
        58000,
        76000,
        90000,
        35000,
        67000,
        81000
    ]
}

df = pd.DataFrame(employee)

print(df)

df["Salary_level"] = pd.cut(
    df["Salary"],
    bins = [0, 40000, 70000, 100000],
    labels = ["Low", "Medium", "High"]
)
print(df)

print(df[df["Salary_level"] == "High"])

print(df["Salary_level"].value_counts())

print(df["Salary_level"].value_counts().idxmax())
#idxmax() returns only the first category with the maximum count. 
#In this dataset, Medium and High both have 3 employees, but idxmax() returns only "Medium" because it appears first.
```

## What is qcut()
You don't decide the ranges.

Instead, Pandas divides the data into equal-sized groups.

For example, if there are 12 salaries and you write:

pd.qcut(df["Salary"], q=3)

Pandas automatically creates 3 groups with about 4 employees each, even if the salary ranges are uneven.

This is very useful in:
Customer segmentation, 
Income analysis, 
Machine learning feature engineering

### Unlike cut(), where you define the ranges, qcut() automatically divides the data into equal-sized groups.

```python
import pandas as pd

employee = {
    "Name": [
        "Rahul",
        "Neha",
        "Aman",
        "Priya",
        "Karan",
        "Sneha",
        "Rohan",
        "Aditi"
    ],
    "Salary": [
        28000,
        42000,
        58000,
        76000,
        90000,
        35000,
        67000,
        81000
    ]
}

df = pd.DataFrame(employee)

print(df)

df["Salary_Group"] = pd.qcut(
    df["Salary"],
    q = 4,
    labels = ["Q1", "Q2", "Q3", "Q4"]
)
print(df)
```

### What does q=4 mean?

It creates 4 equal-sized groups (quartiles).

There are 8 employees.

So Pandas tries to put:

Q1 → 2 employees
Q2 → 2 employees
Q3 → 2 employees
Q4 → 2 employees

Notice:

You didn't specify salary ranges.

Pandas calculated them automatically.

## Difference
### cut()

You decide:

0 --------40000--------70000--------100000
   Low        Medium        High

The number of employees in each group can be very different.

Example:

1. Low      → 2
2. Medium   → 5
3. High     → 1

### qcut()

Pandas decides the ranges.

Example:

Q1 → 2 employees
Q2 → 2 employees
Q3 → 2 employees
Q4 → 2 employees

The salary ranges may not be equal, but the number of employees is approximately equal.

```python
import pandas as pd

employee = {
    "Name": [
        "Rahul",
        "Neha",
        "Aman",
        "Priya",
        "Karan",
        "Sneha",
        "Rohan",
        "Aditi"
    ],
    "Salary": [
        28000,
        42000,
        58000,
        76000,
        90000,
        35000,
        67000,
        81000
    ]
}

df = pd.DataFrame(employee)

print(df)

df["Salary_Group"] = pd.qcut(
    df["Salary"],
    q = 4,
    labels = ["Q1", "Q2", "Q3", "Q4"]
)
print(df)

print(df["Salary_Group"].value_counts())

print(df[df["Salary_Group"] == "Q4"])

# cut() uses user-defined ranges (bins),
# while qcut() automatically creates ranges so that each group contains approximately the same number of observations.
```
